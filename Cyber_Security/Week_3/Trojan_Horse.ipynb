# **What are Trojan Horses?**


what-is-a-computer-trojan.png


                       Image Source: Comodo Enterprise


A Trojan horse is any kind of computer malware that seeks to harm the computing system and often uses the subterfuge of misleading users of the true intentions of what it seeks to do.

The nomenclature takes its roots from the ancient Greek story of a Trojan Horse, which was supposedly meant as a gift to the city of Troy, but actually had hidden Greek soldiers inside the structure, who launched a guerilla attack on the unsuspecting, gullible citizens of Troy, thus winning the war.

# **How Do They Operate?**

A Trojan cannot self-replicate like viruses. It can merely masquerade itself as a legitimate, useful software whilst possessing malicious code and instructions that can disrupt normal files and software. It can be used for a variety of attack vectors. Some of the very common ways people fall prey to it are:



*   One very common methodology is the social engineering mechanism of **phishing**, wherein a user is led to believe that clicking on a particular link recieved as an email attachment will yield some benefit (like winning a lottery), when in reality, it might reek harm.

*   Sometimes, a fake pop-up may arise on web browsers, typically when a user tries to access website without proper **SSL certificates**. This is popularly referred to as **scareware**, and it prompts urgency on the part of the user who can erratically trust the claim and go with it.



*   *Escalation of privilege attack* can be one more method, wherein, hackers exploit the software vulnerabilities in the system to get unauthorized access and they change certain permissions in order to enjoy unlimited privilege.

*   Installing software from non-trustworthy websites.

#**Types of Trojans**

Trojans are categorized based on the types of attacks they can carry out and the precise system of activities that they perform. Some of the well-known ones are enlisted below:



1.   **Backdoor Trojan (Remote Access Trojan)**: As the name suggests, it can open a backdoor to the system being targetted, giving the hacker complete control over it do anything it wants --> be it file manipulation, deletion etc.It can then build a network of bots (aka **botnet**) or a **zombie network** to perpetuate the attack further onto a host of other devices. 


2.  **DDos Trojan**: Following the exploit in the aforementioned method, hackers can install Trojans into a large number of computers and can then commandeer a simultaneous, distributed attack on a target host or web server, thus, depriving others of the resources available and effectively bringing down the host/web-server. This is referred to as a **DDoS attack (Distributed Denial of Service Attack)**. There are a number of ways to do so, the most popular being ***SYN flooding of packets*** or ***repeatedly issuing ICMP echo requests***.



3.  **Proxy Trojan**: The victim's system can be hijacked and transformed into a staging area for further nefarious schemes, but turning into a **proxy server**, part of a botnet. 


4.  **Keylogger Trojan:** It logs each and every keystroke made by the user and compiles the data, which is then sent to the attacker, effectively breaking confidentiality clause of information security.


5. **Trojan Dropper:** A Trojan Dropper can drop different types of standalone malwares (for instance, viruses, worms) to a system. These are typically executable files that contains other files compressed within its body. When activated, these files are decompressed and stored in a temporary location in a computer, from where they can further aggravate and compromise security.

Typically, Trojans cannot be detected even with the aid of antivirus, as they are incredibly good at obfuscation. So that brings up to the question --> how do we detect them? There are some potential signs that may help identify Trojans are probably messing with the system. Some of them are :-

                          


*   Device doesn't respond aptly and is frequently crashing ("***blue screen of death***").

*   On opening the web browsers, there are simply too many pop-up ads and also pop-ups to other websites, which may have mala fide intent. 



*   The system appears to be radically different, with different screen resolution, UI, icons may be missing, some functionalities may be incredibly buggy, etc.

*   Your email folder may witness too many spam mails incoming.

#**How To Safeguard Yourself From These Threats?**

Whilst it is supremely difficult to predict the nature and pattern of threats imposed by Trojans, imbibing good practices can help mitigate any crisis. Some prominent ones are listed as follows :-

*   Keep updating your operating system regularly, as recent versions are more robust, and come with better security patches. 

*   Don't believe in any of the mails that come blindly, as they may instil phishing attacks.



*   Protect accounts with long, complex pattern passwords, which are difficult to crack using password crackers by brute force.

*   Don't click on pop-up ads or links that appear untrustworthy and suspicious and also defer from visiting website which do not have **SSL security certificate clearance** (this can be inferred by looking at the domain's header, which if it is HTTPS, implies that it is secure).



*   Leveraging firewall for filtering out packets entering from Internet to our systems is a good idea. Nowadays, most OS come up with built-in firewall systems to block malicious data. In addition, we can take this up a notch higher by enforcing a **hardware firewall**.
