# **Creating Grids**

###**What are Grids in Figma**

In design, a grid may help with organisation and uniformity, readability, and emphasis and hierarchy. You may establish structure, hierarchy, and rhythm in your design by carefully designing a grid that is appropriate for your content. Grid scaffolding, when done correctly, may remove the guesswork from many elements of your workflow. Grids of various sorts can also benefit in the rationalization of type scales, placement, size, and spacing.

Basically, Layout Grids in Figma helps the users to align objects within a Frame. 
It is easier to give the designs a visual structure with the use of Layout Grids.
They help us in keeping our designs rational and consistent across platforms and devices.

The pixel grid isn't required for layout grids. This implies they aren't constrained by a certain resolution or size.

Grids for layout allow you to:

* Ensure that your branding is consistent across all platforms.

* When creating layouts, make fewer choices.

* Reduce the time it takes to create mock-ups or wireframes by reducing the time required to specify layouts.

* Support a variety of layout methods, including galleries, icons, and full-page layouts.

### **Applying layout grids**

Layout grids may only be applied to frames, which means they can be applied to any top-level frames scaled for a device (desktop, mobile, tablet), nested frames inside your design, or even frames within your components.Â 

*Keep in mind that components are also frames*.

The layout grid settings are located in the right sidebar

To use a layout grid, follow these steps:



1. Select the canvas frame or the Layers panel in the left sidebar



2. In the right sidebar, click + next to Layout grids



3. By default, a uniform grid will be applied to the frame



4. Then click the button next to Grid with the nine dots in a square shape to view the layout grid options and make any necessary changes

### **Types of grids**

A uniform square grid, columns, and rows are the three types of layout grids available.

Uniform grids generate a set of equally spaced fields across your frame (think of a sheet grid paper) at whatever size you choose. Uniform grids allow you to customize the grid's size and color. Square grids are ideal for instances when accuracy is required, such as when drawing symbols or icons.



You may customize the width and height of the grid, as well as the gutter and margins, using column and row grids. They're great for creating web and mobile responsive interfaces.



Grids can be used alone or in combination to provide more sophisticated layouts.

### **Properties of a grid**

The characteristics of your grid may be defined as follows:

The **Count** specifies the number of columns or rows in the grid.

The **Gutter** is the line that separates each column or row.

The **Margin** is the space between the beginning of the column or row and the edge. Only Stretch grids are affected.

The **Offset** specifies how far the column or row starts from the top or left. This is true for rows that are set to the top and columns that are set to the left.

**Fixed** and **stretchy** grids are the two types of layout grids.

You may set the Width (column width) and Height (row height) of **Fixed layout grids**, as well as the Count (number) of columns and rows.



The grid for the layout may be set to:



* For Rows, the top or centre of the frame

* Columns should be placed to the left or centre of the frame.

**Stretchy grids** will adapt to the size of the frame. This allows your designs to respond accordingly when the frame is resized.

The Width or Height of the grid will be automatically determined, based on the frame size.

A layout grid's default **colour** is red (#FF0000), and it has an opacity of 10%.

Any layout grid can have its colour and opacity changed:



1. In the right sidebar, click the button with the nine dots in a square shape.

2. To use the colour swatch, simply click on it:

3. Pick a new colour using the colour picker.

4. Adjust the opacity if needed.

### **Combine layout grids**

You may use many layout grids on a single frame. When creating more complicated Layouts, this comes in helpful.

For example, if you want additional flexibility over vertical and horizontal alignment, you may use both Column and Row layout grids.

1. Choose the frame to which you want to add another layout grid.

2. In the right sidebar, click next to Layout grids.

3. Choose a layout grid from the dropdown menu.

4. Define any extra characteristics that are appropriate.

5. Add more layout grids to the frame by repeating the process.
