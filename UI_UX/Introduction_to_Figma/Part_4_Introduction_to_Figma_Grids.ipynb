## Layout Grids

Some basic terminologies to get started with layout grids are: 
>* Columns: The vertical spatial zones of the screen where content is arranged.
* Rows: The horizontal spatial zones  of the screen that extends from left to right.
* Gutters: The spaces between columns and rows that separates columns and rows.
* Margins: The spacing that separates content and in the far left and right sections of the screen.

## Understanding Grids
>
*	Layout grid is compatible for **frames** of any screen resolution. A grid can also be set on any component with frames included. Grids simplify responsive web design i.e. it works great on desktop, tablet and mobile frames and helps easy switching between them. 
*	**Nested grids** are a powerful feature of figma grids. In simple words, it gives you the functionality to add multiple grids to the same frame for differnet parts of your design. Switching between grids can be done by toggling the visibility. Grids can be differentiated by specifying your choice of **shades and opacities**. 
*	Modifications in the grid parameters are simplified and you can easily use the same grid pattern across multiple frames by simply group selecting the grids and pasting it onto your desired frame.



Layout Grids are of three kinds:
>1. Uniform Grids
2. Column Grids
3. Baseline Grids

Figma grids gives you a great flexibility to improvise upon your designs.

####Type
Type parameter of column are Stretch, Left and Center, while row are Stretch, Top, Bottom and Center.

>
* Stretch: It automatically grow or shink the width of the column/row when you resize the frame.

* Center: It follows a centered gridline. It comes in handy for ultra wide screen displays where much of the horizontal spacing may not be utilized.
* Left: It shifts the layout grid to the leftmost edge of the frame.
*Top: It sets the row guides to the top of the design frame.
*Bottom: It sets the row guides to the bottom of the design frame.


####Offset

It is used to adjust space at the begining of the grid(i.e. top or left).


####Count
The number of columns/rows can be chosen based on the precision of the design.

##Nudge Amount

Setting the right nudge amount according to the grid is required to enable proper aligning.

* Click on the menu.
* Go to Preferences 
* Then click on Nudge amount
* Enter your desired Small nudge and Big nudge details.

Big nudge: Adjusting the big nudge varies the amount of gridlines will shift upon using shift+direction keys. For example, in 8pt grids the approriate big nudge amount is 8.

Small nudge: Adjusting the small nudge varies the amount of gridlines the elements will shift while using the direction keys alone.

##Additional Tips
