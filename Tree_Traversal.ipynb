# Tree Traversal
Printing Each and every element of a tree is call tree traversing.
It can be performed my following ways

                                    Tree Traversal
                                           |
                   |----------------------------------------------|            
           Depth First Traversal                         Bredth First Traversal
                   |

## Depth First Traversal
In DFS we go to one side of the node and traverse it completely and then wo go to another side of the node and traverse it.

### Preorder Traversal
In Preorder Traversal first the node element is printed then the left subtree of the node element and then the elements of right subtree

Tree-Traversals-Preorder.png

Output : 1 2 4 5 3 6 7






```
 void printPreorder(Node node)
    {
        if (node == null)
            return;
 
        /* first print data of node */
        System.out.print(node.key + " ");
 
        /* then recur on left sutree */
        printPreorder(node.left);
 
        /* now recur on right subtree */
        printPreorder(node.right);
    }
```



### Inorder Traversal 
In Inorder traversal first the elements left side of the node are printed then the node element itself and at last elements of right side of nodes are printed.

Tree-Traversals-Preorder.png

Output : 4 2 5 1 6 3 7



```
 void Inorder(Node node)
    {
        if (node == null)
            return;
 
        /* first recur on left child */
        printInorder(node.left);
 
        /* then print the data of node */
        System.out.print(node.key + " ");
 
        /* now recur on right child */
        printInorder(node.right);
    }
```



### Postorder Traversal
In Postorder traversal first the right subtree is printed then the left subtree and at last the node element.


Tree-Traversals-Preorder.png

Output : 7 6 3 5 4 2 1



```
void printPostorder(Node node)
    {
        if (node == null)
            return;
 
        // first recur on left subtree
        printPostorder(node.left);
 
        // then recur on right subtree
        printPostorder(node.right);
 
        // now deal with the node
        System.out.print(node.key + " ");
    }
```
