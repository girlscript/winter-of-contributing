```python
Documentation for HTML <head> tag:-

The <head> tag contains all the metadata, metadata means the data about the HTML document.
This tag contains the <style>,<title>,<favicon>,<meta>,<link>,<script> & <base> tags.
Metadata typically define the document title, character set, styles, scripts and other meta information about the HTML docuement.

The most important part about the <head> tag is that whatever it contains doesn't get displayed on the website. Only the contents of <body> tag get displayed on the website.

Now looking at the different tags that are contained in the <head> tag:-

1. <title> - This tag contains the title of the HTML document. Title is a compulsory element for any HTML document.
             Title must be text-only, and it is shown in the browser's title bar or in the page's tab.
             The title is very much important for SEO(search engine optimization).

             The <title> element:-
             -defines a title in the browser toolbar
             -provides a title for the page when it is added to favorites
             -displays a title for the page in search engine-results

             So a meaningful and accurate title should be chosen.

        Example:-
        <!DOCTYPE html>
        <html> 
        <head> <!--beginning of the head tag-->
  
          <title>Page Title</title> <!-- title tag containing the title of the document/page-->
  
        </head> <!--ending of the head tag-->
  
        <body> <!--beginning of the body tag-->

        Content of the document

        </body> <!--ending of the body tag-->
        </html>

2. <style> -This tag is used to define the styling for that individual HTML page.
            Whatever style is defined in an HTML document's <head> is only applicable for that page.

      Example:-

      <style> <!--beginning of the style tag-->

        body {background-color: powderblue;}
        h1 {color: red;}
        p {color: blue;}

      </style> <!--ending of the style tag-->


3. <link> - This tag is used to link any external file to the current HTML document.
            This tag defines the relationship between the current document and an external resource.
            Mostly used to link CSS files.
          
      Example:-

      <link rel="stylesheet" href="style.css">

      Here rel="stylesheet" defines the relationship of the external resource with the current HTML document i.e the resource is a stylesheet.

      href="style.css" depicts the link of the file using its name and location. Here style.css is located in the same folder as the HTML document so its url is written like that.


4.<meta>- This tag is used to specify the character set, page description, keywords, author of the document, and viewport settings.
          The metadata is not displayed on the page, but is used by browsers (how to display content or reload page), by search engines (keywords), and other web services.

        <meta charset="UTF-8"> <!-- Defines the character set -->
        <meta name="keywords" content="HTML, CSS, JavaScript"> <!--Define keywords for search engines -->
        <meta name="description" content="Free Web tutorials"> <!--Define a description of the web page -->
        <meta name="author" content="Soubhagya Pratik"> <!--Define the author of a page-->
        <meta http-equiv="refresh" content="30"> <!--Refresh document every 30 seconds-->
        <meta name="viewport" content="width=device-width, initial-scale=1.0"> <!-- Setting the viewport to make your website look good on all devices -->

    Example:-

    <!DOCTYPE html>
    <html>
    <head>
    <meta charset="UTF-8">
    <meta name="description" content="Free Web tutorials">
    <meta name="keywords" content="HTML, CSS, JavaScript">
    <meta name="author" content="Soubhagya Pratik">
    </head>

4. <script> - This tag is used to define client-side JavaScripts(JS for frontend not for backend).

        Example:-

        <script>
        function myFunction() {
          document.getElementById("demo").innerHTML = "Hello JavaScript!";
        }
        </script>

        This <script> writes "Hello JavaScript!" into an HTML element with id="demo".

5.<base> - This tag is used to specify the base URL and/or target for all relative URLs in a page.
           The <base> tag must have either an href or a target attribute present, or both.
           There can only be one single <base> tag in a document.

    Example:-

    <head>
    <base href="https://girlscript.tech/" target="_blank">
    </head>

    <body>
    <img src="images/stickman.gif" width="24" height="39" alt="Stickman">
    <a href="tags/tag_base.asp">HTML base Tag</a>
    </body>


Summary of the topics covered:-

-The <head> element is a container for metadata (data about data) and is placed between the <html> tag and the <body> tag.

-The <title> element is required and it defines the title of the document.

-The <style> element is used to define style information for a single document.

-The <link> tag is most often used to link to external style sheets.

-The <meta> element is typically used to specify the character set, page description, keywords, author of the document, and viewport settings.

-The <script> element is used to define client-side JavaScripts.

-The <base> element specifies the base URL and/or target for all relative URLs in a page.
```
