#Classes & ID's

ID’s and Classes are attributes that form the basic fundamentals of HTML. They are an important part of segmenting content and allowing the access to those sections for styling via CSS and adding functionality and interactivity via Javascript. The major difference between the two selectors is the number of times it can be used within a page.

##ID Selectors


- Primarily Ids are used to identify unique elements i.e. that exists only once in the page. Multiple usage of the same id attribute might not give an error as HTML and CSS are designed to be fault tolerant, but it’s a best to keep id attributes unique to serve its purpose.

    ```<section id="home"></section>```
>For example, heading tags of different web sections.


-	Another functionality of ids are to act as  fragment identifier within anchor tags as the value of the href attribute by which elements targeted by ID’s can be used for intra-page navigation. 

    ``` <li><a href="#home">HOME</a></li>```

    In the above example, the anchor tag can be used to navigate to the home section using its id name.


-	This is where multiple usage of the same id attribute will lead to errors, as the inter-page navigation can’t direct to two different locations at the same time.

##Class Selectors


Class selectors are used to target multiple items at the same time with a common feature; i.e. repeating web elements items used for representing fragment similarities. In the code snippet given below, multiple div's with the same class name "container" can be accessed via CSS and Javascript simultaneously.

   ```<div class="container"></div>```

A single section or container in HTML can have more than one classes referenced to it, to add multiple styles at the same time and to improve functionality using javascript.


   ```<button class="button btn-primary">SUBMIT</button>```

CODE SNIPPET:
  ```<style>
    .button {
      all: unset;
      width: 100px;
      height: 50px;
      padding: 0.25rem;

    }
    .btn-primary {
      background-color: #006eff;
      color: white;
    }
  </style>
  <body>
    <button class="button btn-primary">SUBMIT</button>
  </body>
  ```

##Adding styles to ID’s and Classes

-	Id’s can be accessed in CSS by adding a “#” before the id name like “#IdName”.

      In the below example, #home is used to access the home section and add specific styling to it.  

CODE SNIPPET:

``` 
      <style>
          #home
          {
            max-width:500px;
            border: 2px solid red;
          }
      </style>
      <body>
        <section id="home">
            <p>Id selectors identify unique individual elements.</p>
        </section>
      </body>
  ```
  OUTPUT:

-	Classes are accessed with a “.” along with the class name like “.className”.

      In the below example, .container is used to access all div's with class name blueFont.  

CODE SNIPPET:

``` 
      <style>
          .blueFont
          {
            color:blue;
          }
      </style>
      <body>
         <h2 class="blueFont">About Classes</h2>
         <p class="blueFont">Class selectors can be used multiple times</p>
     </body>
```
OUTPUT:

image.png

>Multiple classes can be accessed simultaneously using the css style tags for applying common styling by separating them with comma’s.

CODE SNIPPET:
``` 
  <style>
    .home_text,
    .pricing_text,
    .service_text {
      background: #222222;
      color: white;
      border: 2px solid blue;
      height: 10vh;
      width: 20vw;
      margin:10px;
    }
  </style>
  <body>
    <div class="home_text">
      <h1>HOME</h1>
    </div>
    <div class="pricing_text">
      <h1>PRICING</h1>
    </div>
    <div class="service_text">
      <h1>SEVICE</h1>
    </div>
  </body>
```

OUTPUT:

image.png

 ```const homeSection=document.querySelector("#home")```
