<h1>Introduction<h1>

Hello everyone, toady we will be discussing about webview in android using java. So let's see what's this webview is, android webview is used to display webpages in your android applications.
The webview is basically like a framework which allows us to specify viewport and style properties that make your web pages appear at the proper size and scale on all screen configurations for all major web browsers. We can even define an interface between your Android app and our web pages that allows javaScript in the web pages to call upon APIs in your app for providing Android APIs to our web-based application.


In order to add WebView to your application, you have to add < WebView > element to your xml layout file.<br>

Syntax :<br>

          < WebView  xmlns:android="http://schemas.android.com/apk/res/android"<br>
          android:id="@+id/webview"<br>
          android:layout_width="fill_parent"<br>
          android:layout_height="fill_parent"
        / >



After adding this in our xml file, in order to use we need to reference this view in our java file. We will be creating the object of this class -<br>
Syntax :<br>
                  
      WebView webview = (WebView) findViewById(R.id.webview);

**Note :** However, you shouldn't develop an Android app simply as a means to view your website. Rather, the web pages you embed in your app should be designed especially for that environment.

The web page can be loaded from same application or URL. Android WebView uses webkit engine to display web page.

The android.webkit.WebView is the subclass of AbsoluteLayout class.

<h3>Diiferent methods provided by WebView</h3>

Apart from just loading url there are several other method's also which are provided by webview to have more control over webview by using webview class.

1.   canGoBack() - This method specifies the WebView has a back history item.

1.   canGoForward() - This method specifies the WebView has a forward history item.

1.   clearHistory() - This method will clear the WebView forward and backward history.

1.   destroy() - This method destroy the internal state of WebView.

1.   findAllAsync(String find) - This method find all instances of string and highlight them.
2.   getProgress() - This method gets the progress of the current page.

2.   getTitle() - This method return the title of the current page.

2.   getUrl() - This method return the url of the current page.
