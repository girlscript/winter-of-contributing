## **What is the use of listener in android?**

Android Listeners are **used to capturing events**. When, for instance, the user interacts with the Android system by clicking on a button, the Listeners would prompt the underlying activity to do the task associated with the button click.

## **Basic Event Listeners in android:**

### **View Event Listeners:**

Any View (Button, TextView, etc) has many event listeners that can be attached using the setOnEvent pattern which involves passing a class that implements a particular event interface. The listeners available to any View include:

*  <a href="https://stackoverflow.com/questions/25803727/android-setonclicklistener-method-how-does-it-work">setOnclickListener</a> - Callback when the view is clicked
* <a href="https://developer.android.com/reference/android/view/View.OnDragListener#:~:text=Interface%20definition%20for%20a%20callback,'false'%20from%20this%20callback.">setOnDragListener</a> - Callback when the view is dragged
* <a href="https://www.tabnine.com/code/java/methods/android.widget.TextView/setOnFocusChangeListener">setOnFocusChangeListener</a> - Callback when the view changes focus
* <a href="https://developer.android.com/reference/android/view/View.OnGenericMotionListener">setOnGenericMotionListener</a> - Callback for arbitrary gestures
* <a href="https://developer.android.com/reference/android/view/View.OnHoverListener">setOnHoverListener</a> - Callback for hovering over the view
* <a href="https://www.tabnine.com/code/java/methods/android.view.View/setOnKeyListener">setOnKeyListener</a> - Callback for pressing a hardware key when view has focus
* <a href="https://developer.android.com/reference/android/view/View.OnLongClickListener">setOnLongClickListener</a> - Callback for pressing and holding a view
* <a href="https://stackoverflow.com/questions/11690504/how-to-use-view-ontouchlistener-instead-of-onclick">setOnTouchListener</a>   - Callback for touching down or up on a view

### **AdapterView Event Listeners:**

In addition to the standard View listeners, AdapterView descendants have a few more key event listeners having to do with their items:

* <a href="https://stackoverflow.com/questions/4709870/setonitemclicklistener-on-custom-listview">setOnItemClickListener</a> - Callback when an item contained is clicked
* <a href="https://www.tabnine.com/code/java/methods/android.widget.ListView/setOnItemLongClickListener">setOnItemLongClickListener</a> - Callback when an item contained is clicked and held
* <a href="https://www.tabnine.com/code/java/methods/android.widget.Spinner/setOnItemSelectedListener">setOnItemSelectedListener</a> - Callback when an item is selected

### **EditText Common Listeners:**

In addition to the listeners described above, there are a few other common listeners for input fields in particular.

* <a href="https://stackoverflow.com/questions/43493825/how-to-implement-on-one-method-of-textwatcher-in-android">addTextChangedListener</a> - Fires each time the text in the field is being changed
* <a href="https://stackoverflow.com/questions/16708146/android-setoneditoractionlistener-doesnt-fire">setOnEditorActionListener</a> - Fires when an "action" button on the soft keyboard is pressed

### **onTouchListener:**

The user interacts with the application while touching the screen in a different location for his desired action. Android applications understand the touch and process this touch and take appropriate action. 
We can customize this action as per our requirement, with the help of onTouchListener.

OnTouchListener is used *to capture a touch event and execute the task associated with it.*

Here we are going to create an onTouchListener on our ConstraintLayout
and going to see how we can move our view around the screen on our application. 

Steps 1 and 2 will remain the same.

 **Step 3:** Give an id to your ConstraintLayout

```python
<?xml version="1.0" encoding="utf-8"?>
<androidx.constraintlayout.widget.ConstraintLayout xmlns:android="http://schemas.android.com/apk/res/android"
    xmlns:app="http://schemas.android.com/apk/res-auto"
    xmlns:tools="http://schemas.android.com/tools"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    android:id="@+id/parent_layout"
    tools:context=".MainActivity">
```

 **Step 4**: Find the id into your onCreate method.

**Step 5**: Set onTouchListener on the layout.

```python
package com.sb2318.myapplication;

import androidx.appcompat.app.AppCompatActivity;
import androidx.constraintlayout.widget.ConstraintLayout;


import android.annotation.SuppressLint;
import android.os.Bundle;
import android.view.MotionEvent;
import android.view.View;
import android.widget.Button;


import android.widget.Toast;

public class MainActivity extends AppCompatActivity {

     private Button button;
     private ConstraintLayout myLayout ;
    private float x;
    private float y;
    private int i=0;

    @SuppressLint("ClickableViewAccessibility")
    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_main);

        button= findViewById(R.id.button);
        // step 4
        myLayout= findViewById(R.id.parent_layout);


        button.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                Toast.makeText(MainActivity.this,"Button is clicked",Toast.LENGTH_LONG).show();
            }
        });

            // step 5
        myLayout.setOnTouchListener(new View.OnTouchListener() {

            @SuppressLint("ClickableViewAccessibility")
            @Override
            public boolean onTouch(View v, MotionEvent event) {
                 x= event.getX();
                 y= event.getY();


                     button.setX(x);
                     button.setY(y);
                     button.setText("You bring me here");

                 if(++i<60)
                     return true;
                 else
                     return false;
            }
        });
    }
}
```
