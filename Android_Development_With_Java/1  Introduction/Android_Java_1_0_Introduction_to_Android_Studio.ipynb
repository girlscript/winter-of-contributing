### **How can we use Android Studio to develop an application?**

Let's create an application that will take two Integer inputs from the user. Then with the click of a button, we see the output, which is the sum of those two integers.

For this, we will first create a project, as we did a while ago. I have named my project here **Add-on**, you are independent in this regard.
An application usually consists of two parts. One is the user interface, which is visible in front of the user. Another is its internal coding part, which enlivens our user interface.


As soon as your project opens, you will see two files. One is MainActivity.java, where we will do internal coding, Another is activity_main.xml, where we will design the user interface.

Let us first create the user interface.

**Required view :**
*   Two <a href ="https://www.tutlane.com/tutorial/android/android-edittext-with-examples">EditText</a> (For user input.)

*   One <a href="https://www.javatpoint.com/android-working-with-button#:~:text=Android%20Button%20Example%20with%20Listener&text=If%20user%20clicks%20button%2C%20sum,button%20in%20activity's%20xml%20file.">Button</a> (To determine the result.)

*   One <a href="https://abhiandroid.com/ui/textview">TextView</a> (To show the result.)

Here I have used ConstraintLayout as the parent layout.

<a href="https://www.journaldev.com/13590/android-constraintlayout"> See an example of the use of ConstraintLayout.</a>





```python


<?xml version="1.0" encoding="utf-8"?>
<androidx.constraintlayout.widget.ConstraintLayout xmlns:android="http://schemas.android.com/apk/res/android"
    xmlns:app="http://schemas.android.com/apk/res-auto"
    xmlns:tools="http://schemas.android.com/tools"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    tools:context=".MainActivity">

    <EditText
        android:id="@+id/first_integer"
        android:layout_width="0dp"
        android:layout_height="wrap_content"
        android:ems="10"
        android:hint="First Integer"
        android:inputType="textPersonName"
        android:minHeight="48dp"
        app:layout_constraintBottom_toBottomOf="parent"
        app:layout_constraintEnd_toEndOf="parent"
        app:layout_constraintHorizontal_bias="0.0"
        app:layout_constraintStart_toStartOf="parent"
        app:layout_constraintTop_toTopOf="parent"
        app:layout_constraintVertical_bias="0.228" />

    <EditText
        android:id="@+id/second_integer"
        android:layout_width="0dp"
        android:layout_height="wrap_content"
        android:layout_marginTop="16dp"
        android:ems="10"
        android:hint="Second Integer"
        android:inputType="textPersonName"
        android:minHeight="48dp"
        app:layout_constraintEnd_toEndOf="parent"
        app:layout_constraintStart_toStartOf="parent"
        app:layout_constraintTop_toBottomOf="@+id/first_integer" />

    <Button
        android:id="@+id/button"
        android:layout_width="wrap_content"
        android:layout_height="wrap_content"
        android:layout_marginTop="24dp"
        android:text="See result"
        android:textSize="18sp"
        app:layout_constraintEnd_toEndOf="parent"
        app:layout_constraintStart_toStartOf="parent"
        app:layout_constraintTop_toBottomOf="@+id/second_integer" />

    <TextView
        android:id="@+id/result"
        android:layout_width="0dp"
        android:layout_height="wrap_content"
        android:gravity="center"
        android:text="TextView"
        android:textSize="28sp"
        app:layout_constraintBottom_toBottomOf="parent"
        app:layout_constraintEnd_toEndOf="parent"
        app:layout_constraintHorizontal_bias="0.0"
        app:layout_constraintStart_toStartOf="parent"
        app:layout_constraintTop_toBottomOf="@+id/button"
        app:layout_constraintVertical_bias="0.255" />
</androidx.constraintlayout.widget.ConstraintLayout>



```

Now let's look at the coding part, Here we will clarify our concept step by step.

First, let's see which views we want to set the action. In this case, except for the parent layout, we will perform some action on each View.

The second step is that we will give a unique id to each identified View.

Now we are ready for the coding part.

In the next step, At MainActivity class, we will declare each required view, and find their id inside the onCreate method.

We have already discussed which actions we want to perform in which views.

Now it's time to look at the coding part.

```python


package com.application.addon;

import androidx.appcompat.app.AppCompatActivity;

import android.os.Bundle;
import android.view.View;
import android.widget.Button;
import android.widget.EditText;
import android.widget.TextView;
import android.widget.Toast;

public class MainActivity extends AppCompatActivity {

    // declare the required view
      private EditText firstInteger, secondInteger;
      private Button button;
      private TextView result;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_main);

        //  find the view
        firstInteger= findViewById(R.id.first_integer);
        secondInteger= findViewById(R.id.second_integer);
        button= findViewById(R.id.button);
        result= findViewById(R.id.result);

        // set action on the button
        button.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {

                // check whether the edit text field are empty or not.
                if (firstInteger.getText().toString().isEmpty() || secondInteger.getText().toString().isEmpty()) {

                    // show a toast
                    Toast.makeText(MainActivity.this, "Input field cannot be empty", Toast.LENGTH_SHORT).show();
                } else {


                    // get the user input from EditText
                    int first = Integer.parseInt(firstInteger.getText().toString());
                    int second = Integer.parseInt(secondInteger.getText().toString());

                    int sum = first + second;
                    result.setText(String.valueOf(sum));
                }

            }
        });
    }
}


```
