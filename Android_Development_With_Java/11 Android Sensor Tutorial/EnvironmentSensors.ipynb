# Environmental sensors:

---



These sensors measure the various environmental parameters such as ambient air temperature and pressure, illumination, and humidity.

These sensors measure the various environmental parameters such as ambient air temperature and pressure, illumination, and humidity.
This category includes Barometers(Pressure), Photometers(light intensity), and Thermometers(Temperature).

The Android platform provides four sensors that let you monitor various environmental properties. All four environment sensors are hardware-based and are available only if a device manufacturer has built them into a device. Except for the light sensor, which most device manufacturers use to control screen brightness, environment sensors are not always available on devices. Because of this, you must verify at runtime whether an environment sensor exists before you attempt to acquire data from it.

**Environmental sensors that are supported on the Android platform:**



1.   <a href="https://developer.android.com/reference/android/hardware/Sensor#TYPE_AMBIENT_TEMPERATURE">**TYPE_AMBIENT_TEMPERATURE**</a> (Ambient air temperature.(°C))

2.   <a href="https://developer.android.com/reference/android/hardware/Sensor#TYPE_LIGHT">**TYPE_LIGHT**</a> (Illuminance (lx))
3.   <a href="https://developer.android.com/reference/android/hardware/Sensor#TYPE_PRESSURE">**TYPE_PRESSURE**</a> (Ambient air pressure (hPa or mbar))
4.   <a href="https://developer.android.com/reference/android/hardware/Sensor#TYPE_RELATIVE_HUMIDITY">**TYPE_RELATIVE_HUMIDITY**</a> (Ambient relative humidity (%))
5. <a href="https://developer.android.com/reference/android/hardware/Sensor#TYPE_TEMPERATURE">**TYPE_TEMPERATURE**</a> (Device temperature (°C))

# Monitoring sensor events:


---



pic.png

Let us assume this is our Android system and our application. This android system has many built-in sensors we want to get the value of one particular sensor in our application. To do that we need to follow some steps.

**Step 1:** Implement <a href="https://developer.android.com/reference/android/hardware/SensorEventListener">**SensorEventListener**</a> interface.

**Step 2:** Invoke <a href="https://developer.android.com/reference/android/hardware/SensorEventListener#onSensorChanged(android.hardware.SensorEvent)">**onSensorChanged()**</a> and <a href="https://developer.android.com/reference/android/hardware/SensorEventListener#onAccuracyChanged(android.hardware.Sensor,%20int)">**onAccuracyChanged()**</a> methods.

**Step 3:** Register listener in <a href="https://developer.android.com/guide/components/activities/activity-lifecycle#onresume">**onResume()**</a> method.

**Step 4:** Unregister listener in <a href="https://developer.android.com/guide/components/activities/activity-lifecycle#onpause">**onPause()**</a> method.

We need to follow this four steps to get the value of any sensor in our application.

To know more about registering and unregistering a sensor,<a href="https://support.rhombussystems.com/hc/en-us/articles/360038529792-Registering-or-Unregistering-a-Sensor">**click here**</a>

 

**In this example, we are going to use Light sensors,
In this way, we will also look at the implementation of the light sensor.**

```python
package com.sb2318.myapplication;

import androidx.appcompat.app.AppCompatActivity;
import androidx.constraintlayout.widget.ConstraintLayout;


import android.annotation.SuppressLint;
import android.content.Context;
import android.content.Intent;
import android.content.IntentFilter;
import android.hardware.Sensor;
import android.hardware.SensorEvent;
import android.hardware.SensorEventListener;
import android.hardware.SensorManager;
import android.os.Bundle;
import android.view.MotionEvent;
import android.view.View;
import android.widget.Button;


import android.widget.TextView;
import android.widget.Toast;

import java.util.List;

// step 1 : implement SensorEventListener interface and override the method
public class MainActivity extends AppCompatActivity implements SensorEventListener  {

    // step 2 :Create an object of TextView and SensorManager and float object
    private TextView textView;
    private SensorManager sensorManager;
    private Float changedValue;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_main);

       // step 3 : Initialize all the variable
        textView= findViewById(R.id.textView);
        sensorManager= (SensorManager) getSystemService(Context.SENSOR_SERVICE);
        
    }


    @Override
   public void onSensorChanged(SensorEvent event) {

        // here we will write the code to get the value of sensor
        // step 4: get the value
        changedValue= event.values[0];
        // step 5: show the value
        textView.setText(String.valueOf(changedValue)); // whatever value we will get from the sensor event


    }

    @Override
    public void onAccuracyChanged(Sensor sensor, int accuracy) {
     
        // For now, leave it as it is.
    }

    // step 6: register the listener
    @Override
    protected void onResume() {
        super.onResume();

        sensorManager.registerListener(this, sensorManager.getDefaultSensor(Sensor.TYPE_LIGHT),SensorManager.SENSOR_DELAY_NORMAL );

    }

    // step 7: unregister the listener
    @Override
    protected void onPause() {
        super.onPause();
        sensorManager.unregisterListener(this);
    }

   
}
```
