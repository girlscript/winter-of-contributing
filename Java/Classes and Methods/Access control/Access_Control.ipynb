#Access Control in Java
Let’s write a simple class, a bare-bone class:
`class Sample{ }.`

This code will compile but you can also add “modifiers” before declaring the class. These modifiers, as we call it, control the access or scope of any field, method, class, or constructors.
<br><br>

**Access modifiers**

There are four access controls and but only three access modifiers. The fourth access control called default, or package access is the default access control applied on a Java class. Hence, we say out of the four access controls, three of them “modify” the access of the class hence three access modifiers and four access controls.
These four access controls are:
  Public
  Private
  Protected
  Default
We’ll see what job each of these perform. 
<br><br>

**Accessibility**

If one class, say class A has access to class B, then class A  can create an instance of class B, or extend B or access data like methods or variables from class B. This accessibility is what we can control via modifiers.
<br><br>
##Access Control in classes
**Default access**

Default access, as the name suggests is the default access control provided to a Java class. It is a package level access, because a class with default access can only be accessed by classes in that same package. Let us have a look at the following example.
```
package yash;
class Food { }

```
Now let us look at another class in a different package:
```
package raj;
import yash.Food;
class Pizza extends Food { }
```

Here, the import statement will try to import the Food file. Food file will compile properly, but when we compile Pizza, we will get an error. This happens because Food has default access and both classes are in different packages. We can solve this problem with these 2 methods:
1.	Put both classes in same packages
2.	Change the access of Food (superclass) to public
<br><br>

**Public access**

Public, as the name suggests allows the class to be accessed anywhere, even in different packages. But do remember, in order to access the class in a different package, you need the to import it via the import statement. Let us now look at the previous example in the “default access” section with a change.



```
package yash;
public class Food { }
```
Now let us look at another class in a different package:


```
package raj;
import yash.Food;
class Pizza extends Food { }
```

Now Pizza can extend Food without any compilation errors. It can also create an instance of Food and also access any data of Food.
<br><br>

###Nonaccess class modifiers
You can modify a class declaration using keywords final, static, abstract and strictfp.

**Final classes**

Final keyword means the class cannot be inherited OR no subclass can be created from the class. Make a final class only if you need an absolute guarantee that no methods of the class should be overridden. This is the reason why the core java classes are marked final internally. Because they should not be overridden.

Final classes cannot be subclassed even if they are declared public. Let’s see an example:



```
package yash;
public final class Food { }
```
Now let us look at another class in a different package:



```
package raj;
import yash.Food;
class Pizza extends Food { }
```
Pizza file will not compile even if it’s superclass Food is public because a final keyword is added to it.

Final class violate the OOPs principle. So unless there is a safety issue in your project, do not mark a class final.
<br><br>

**Abstract classes**

These classes can never be instantiated. Its mission is to be extended by another class. They are like interfaces. With abstract classes, with abstract classes, you can declare fields that are not static and final, and define public, protected, and private concrete methods. With interfaces, all fields are automatically public, static, and final, and all methods that you declare or define (as default methods) are public. In addition, you can extend only one class, whether or not it is abstract, whereas you can implement any number of interfaces.

##Access control in class members
A class can only be public or default whereas it’s members can have all four: private, public, protected and default.
<br><br>
**Public members**

When a method or variable is declared public, it means that all other classes, including the ones outside the package, can access that method or variable. Let us see an example.


```
package eat;
import foodie.*;
class Burger
{
	public static void main(String[] args)
	{
		Jalapeno j=new Jalapeno();
		j.taste();
  }
}

```

Now let us see another class:

```
package foodie;
public class Jalapeno
{
	public void taste()
	{
		System.out.println(“jalapeno”);
  }
}

```
Burger and Jalapeno are in different packages but Jalapeno can invoke Burger without any problems because both methods are marked public.
<br><br>
**Private members**

Private members cannot be accessed by any code in another class other than the class where it was declared. Let us see an example:

```
package friends;
public class Party
{
	private String goGoa()
  {
	  return “fun”;
  }
}

```
Let us have a look at another class:
```
package friendships;
import friends.*;
class HaveFun
{
	public void tryIt()
  {
	  Party p=new Party(); //so far, it’s great because Party is public
	  System.out.println(p.goGoa()); //error
  }
}

```
goGoa() method is private, so no other class can use it and hence HaveFun() will not compile.

Now you may ask can an inherited class or subclass use a private method of its superclass? The answer is no. Let us have a look at the previous example with changes:

```
package friends;
public class Party
{
	private String goGoa()
  {
	  return “fun”;
  }

}

```
Let us look at a subclass:

```
package friends;
class Chips extends Party
{
	public void tasteChips() 		//so far so good, superclass is public
  {
	  System.out.println(goGoa()); //error
  }
}

```
<br><br>
**Protected**

If methods or variables are declared protected, then they are:
1.	Accessible within the same package
2.	Can be accessed through inheritance by subclasses even in different packages.

Let us look at an example:

```
package foodie;

public class Food
{
protected void show()
	{
		System.out.println("Pizza");
	}
}

Let us look at another class


package gourmet;

import foodie.*;

// Class Italian is subclass of Food
public class Italian extends Food {
	public static void main(String args[]) {
		Italian i = new Italian();
		i.show();
	}

}

```

Here both the methods are in different packages but as show() is protected, Italian, the subclass can invoke show().
<br><br>
**Default**

The only difference between default and protected is that default does not have the superpower to be accessed in different packages in a subclass.

Let us look at an example:

```
package foodie;

public class Food {
	void show() {
		System.out.println("Pizza");
	}
}

```
Let us look at the Italian file now:
```
package gourmet;

import foodie.*;

// Class B is subclass of A
public class Italian extends Food {
	public static void main(String args[]) {
		Italian i = new Italian();
		i.show();
	}

}

```
Italian will not compile as the show() method is default. So it cannot be accessed outside the foodie package.
