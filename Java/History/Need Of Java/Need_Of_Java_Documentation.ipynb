# Why is Java so popular?
The main reason of Java's popularity is **Platform Independence** that means it can run on different type of computers which are compatible with Java Runtime Environment(JRE). Java is considered to be a simple language and they come with wide variety of utility functions which also makes it popular.Java is 26 years old (Java 1.0 was released in 1995), many of the biggest organisations and companies in the world have built their applications with the help of the language.

#Need of Java
As Java is Simple and Easy to use, Object Oriented Programming (OOP) Language, Robust Language i.e, it is capable of handling run-time errors, supports exception handlings, Platform Independent and with security features. <br>Because of these features it is widely used in areas like:
1.   **Web Applications**<br>
Java provides *Servlet* and *JSP* which allows us to make any web application.We can create static website with help of HTML and style it with CSS, but for dynamic wbsite we need server-side technology and here we use Java.  Many government, insurance, healthcare, education departments have built their application with java. Google's Gmail has also used Java.
2.   **Mobile Application**<br>
Java is the most popular programming language used in Mobile Applications because the Mobile OS is written in Java and its API are designed for Java. Java runs on Java Virtual Machine(JVM) and Android uses Dalvik Virtual Machine(DVK) to execute class files, but the code is still in Java. Spotify, Twitter uses Java for Android Mobile Application Development.
3.   **Desktop Application**<br>
Desktop Applications can easily be developed in Java. With help of APIs like *AWT*(Abstract Windowing Toolkit), *Swing*, *JavaFX* to build desktop applications. Swing is a GUI widget toolkit,which provides certain advanced elements. Acrobat Reader is an example of Desktop Application.
4.   **Scientific Application**<br>
Java is often used for scientific applications which includes mathematical operations and natural language processing. We use Java because it is safe,portable and maintainable. MATLAB use Java for interactive user-interface and core part of the system e.i, back-end.
5.   **Enterprise Application**<br>
Enterprise Application is a software system that operates in a corporate organization rather than individual users. Java Enterprise Edition(Java EE) is a popular platform that provide API and run-time environment as well as it is considered as the backbone for variety application. We can make the application easily scalable in order to increase the number of users on the enterprise application sites. Flipkart,Infosys uses Java.
