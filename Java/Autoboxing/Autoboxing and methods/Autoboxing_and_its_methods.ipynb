#Autoboxing

Since Java 5, we have a feature called autoboxing. With this feature, we need not be concerned that the class will return a wrapper class or the primitive data type as Java does the conversion for you.

The Java compiler automatically does the conversion of the primitive class into its corresponding wrapper class. For example  converting an int to an Integer, boolean to Boolean and so on.

Let us understand this with an example.



```
List<Double> amount=new ArrayList<>();
amount.add(1001.23);
```

This will not show error as double is autoboxed to its wrapper class Double automatically.

You could still write that the long way by writing something like `amount.add(new Double(70))` and pass the wrapper object.

The Java compiler applies autoboxing when a primitive value is:

*   Passed as a parameter to a method that expects an object of the corresponding wrapper class.

*   Assigned to a variable of the corresponding wrapper class.

Here's a full program example to demonstrate autoboxing.



```
class Example{  
  public static void main(String args[]){  
    int a=100;  
        Integer a1=new Integer(a);
  
        Integer a2=3;//Boxing  
          
        System.out.println(a1+" "+a2);  
 }   
}  
```
Output:


```
100 3
```

#Unboxing

Unboxing is the exact opposite of autoboxing. Unboxing enables the conversion of wrapper class object into its corresponding primitive.

Let us consider the amount example again.

```
List<Double> amount=new ArrayList<>();
amount.add(1001.23);
double unboxed=amount.get(0);
```

Here the second line autoboxes double into Double as shown before. But now the third line in this code gets the Double and unboxes it into its primitive.

The Java compiler applies unboxing when an object of a wrapper class is:

*   Passed as a parameter to a method that expects a value of the corresponding primitive type.

*   Assigned to a variable of the corresponding primitive type.

Here's a full program example to demonstrate unboxing.

```
class Example{  
  public static void main(String args[]){  
    Integer i=new Integer(100);  
        int a=i;  
          
        System.out.println(a);  
 }   
}
```

Output:

```
100
```

#Special cases

###Unboxing a null



```
List<Double> amount=new ArrayList<>();
amount.add(null);
double unboxed=amount.get(0);
```

In this code, in the second line autoboxing occurs and null is added to the list. In the third line, when we try to access null, it gives us Null Pointer Exception as we are accessing null.

###Being careful while autoboxing integer

Let us take this code for example.


```
List<Integer> amount = new ArrayList<>();
amount.add(1);
amount.add(2);
amount.remove(1);
System.out.println(amount);
```

It actually outputs 1. After adding the two values, the List contains [1, 2]. We then request the element with index 1 be removed. That’s right: index 1. Because there’s already a remove()
method that takes an int parameter, Java calls that method rather than autoboxing. If you want to remove the 2, you can write numbers.remove(new Integer(2)) to force wrapper class use.
