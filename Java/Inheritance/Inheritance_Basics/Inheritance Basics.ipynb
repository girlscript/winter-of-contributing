## Inheritance

Inheritance in Java is a mechanism in which one object acquires all the properties and behaviors of a parent object. It is an important part of OOPs (Object Oriented programming system).

The idea behind inheritance in Java is that you can create new classes that are built upon existing classes. When you inherit from an existing class, you can reuse methods and fields of the parent class. Moreover, you can add new methods and fields in your current class also.

Suppose A is super class of B. Then class B can inherit all the features (Methods and Instance Fields) of class A except the private members.
Super and Sub Classes May Either Belong to Same  or Different Packages


* The class whose properties are extended is known as super or base or parent class.

* The class which  extends the properties of super class is known as sub or derived or child class.

* A class can either extends another class or can implement an interface

Inheritance represents the **IS-A relationship** which is also known as a parent-child relationship.

## Types of Inheritance

### Single-level Inheritance:


When a class inherits another class, it is known as a single inheritance.There is a single child class and a single parent class, as depicted in the image above.Let us understand this with the help of an example.

In the above example we have two classes:
* ProgrammingLanguages is the parent class.
    * This class has a method called message().
* Java Class is the child class.
    * This class has a method called print().
* The child class object can access the methods and variables of the Parent class.This is called Single-level Inheritance

### Multi-level Inheritance:

When there is a chain of inheritance, it is known as multilevel inheritance. As you can see in the example given below, the Child class inherits the Parent class, which inherits the GrandParent Class.So there is a chain of inheritance.This is called as Multilevel Inheritance.

In the above example we have three classes:
* Grandparent Class is the base class.
    * This class has a method called mssg1().
* Parent Class inherits the Grandparent class and is the parent of Child class.
    * This class has a method called mssg2().
* Child Class inherits both the Parent and the Grandaparent Classes.
    * This class has a method called mssg3().
* The child class object can access the methods and variables of the Parent class and the Grandparent class.This is called Multilevel-level Inheritance

### Hierarchical Inheritance:

When two or more classes inherits a single class, it is known as hierarchical inheritance. In the example given below, both Java and Python Class extend the ProgrammingLanguages Class.

In the above example we have three classes:
* ProgrammingLanguages is the parent class.
    * This class has a method called message().
* Java Class is the child class as it inherits the ProgrammingLanguages class.
    * This class has a method called print1().
* Python Class is the child class as it inherits the ProgrammingLanguages class.
    * This class has a method called print2().
* The child class object can access the methods and variables of the parent class.This is called Hierarchical Inheritance
