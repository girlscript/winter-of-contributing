###**MULTIDIMENSIONAL ARRAY**

You are determined to learn Java in and out. So, you walk to a library nearby to get the book named “Java The Complete Reference, 8th Edition”. You ask the librarian where to find the book in that big library. 

book.png

After a quick scan in her database, she tells you that the book is available in the 3rd cell of the 2nd row in the 2nd shelf. So, what do you do? You walk straight to the 2nd shelf, find the 2nd row and locate the 3rd cell. Walla!  you have found the book you wanted!!!

Now let’s figure out how the librarian had track of all the books in the library. For simplicity’s sake, lets imagine this poor library has only one shelf of books. This shelf contains many rows and each row contains many cells (or say columns). So, this shelf is a collection of rows and columns. Similarly in computer terms we term this shelf as an array which is a collection of rows and columns. 

Now consider our original library with many shelves (or arrays). So, we can conclude that this library is a collection of many arrays or an array of arrays. In Java we term this as a MULTIDIMENSIONAL ARRAY. Now how do we code this in Java? 

2d.png

**SYNTAX**

```
dataType [dimension1][dimension2]…..[dimensionN] arrayName = new dataType[size1][size2]…..[sizeN];
```
Let’s look what these keywords mean.
*   dataType
      *   type of data to be stored in the array
      *   Example: int, char, float, etc.
*   arrayName
      *   name of the array
*   dimension1, dimension2, …….., dimensionN
      *   dimension of the array to be created
      *   Example: 
            1.   for a 1D integer array –> int[ ] x = new int[5];
            2.   for a 2D char array –> char[ ][ ] c = new char[2];
*   size1, size2, ……., sizeN
    *  sizes of the corresponding dimension

***Note*** 

As the library shelves contain only books and no other objects (say toys), the Java multidimensional array also contains the same data type for all dimensions i.e., say in a 2D array both the row and column data must be of the same data type.



There is another way to declare a multidimensional array by not specifying the size of the dimensions explicitly.

```
dataType [dimension1][dimension2]…..[dimensionN] arrayName ={
								      {
								         {valueD1R1C1, valueD1R1C2,…..},
								         {valueD1R2C1, valueD1R2C2,…..},
								          ………………………..								                          
                                      }
								      {
								         {valueD2R1C1, valueD2R1C2,…..},
								         {valueD2R2C1, valueD2R2C2,…..},
								          ………………………..								                    
                                      }
								      …………………………..
								      {
								         {valueDnR1C1, valueDnR1C2,…..},
								         {valueDnR2C1, valueDnR2C2,…..},
								          ………………………..								                    
                                      }
    };

```
Example: 

int[ ][ ] x = { {1, 2}, {3, 4} };

int[ ][ ][ ] y = { { {0, 1}, {2, 3} }, { {4, 5}, {6, 7} } };

Now let’s declare a 2D array for our library. In our case the data type is String. Let’s name the array as books. And say we have a size of 2 rows and 3 columns. So, we have:

							String[][] books = new String[2][3];

We can find the number of rows in the array by

					books.length
We can find the number of columns in the row by

					books[row].length

**ACCSESSING AN ELEMENT**

Now recall how did the librarian tell you the location of the book. She used certain figures (or indices) to tell you go there, look there. The same way to access an element from a multidimensional array we use indices. 

Example: 

In a 2D array – x[0][1] accesses the element in row 0 and column 1

In a 3D array – y[2][3][1] accesses the element present in the second array, 3rd row and 1st column

In our library example, we access an element by specifying the row and column number like 

`books[row][col]`


**ASSIGNING VALUES**

Our 2D books array has no real elements in it. So now we’ll insert some books into the array. In Java we have two ways to assign values. One is by providing a value to each cell by accessing it individually. 

```
books[0][0] = AliceInWonderLand;

books[0][1] = TheThreeMusketeers;

books[0][2] = TheThreeLittlePigs;

books[1][0] = DataCommunication;

books[1][1] = DatabaseSystems;

books[1][2] = JavaTheCompleteReference;
```
This is an exhaustive approach. For large arrays it is impossible to assign every element in this way.

The other way is an iterative approach. Input can be given by the user and that element will be assigned to its corresponding position.

```
Scanner sc = new Scanner(System.in);
for (int row = 0; row < 2; row++){
	for (int col = 0; col < 3; col++){
		System.out.print("Book at row" + (row+1) + " col " + (col+1) + ": ");
		        books[row][col] = sc.nextLine();
	    }
}	
```


**DISPLAYING VALUES**

We can easily display the array by using two for loops. Individual elements can be accessed and displayed.

```
for (int row = 0; row < 2; row++){
	for (int col = 0; col < 3; col++){
		System.out.print(books[row][col] + "\t");
	}
	System.out.println();
}
```

***Note:***

The tab space and the next line which I have included in the print statements are simply for alignment, to make it look beautiful by displaying it in a tabular form. You can simply print the elements one by one.

We’ve come a long way. So now let’s join all the snippets of code that we have coded for our library  and see how our book collection looks like.


````python
import java.util.Scanner;		//Import util package to use Scanner class to get input from the user

public class Main
{
	public static void main(String[] args) {
	    	Scanner sc = new Scanner(System.in);	
		String[][] books = new String[2][3];	//Declaration of 2D array
		for (int row = 0; row < 2; row++){
		    for (int col = 0; col < 3; col++){
		        System.out.print("Book at row" + (row+1) + " col " + (col+1) + ": ");
		        books[row][col] = sc.nextLine();	//Get input from the user and assign it to the corresponding indices
		    }
		}
		System.out.println("List of Books");
		for (int row = 0; row < 2; row++){
		    for (int col = 0; col < 3; col++){
		        System.out.print(books[row][col] + "\t");	//Display the elements present in our 2Darray
		    }
		//To print the 2D array in a tabular format, go to the next line after each row
		    System.out.println();	
		}
	}
}
````

*Walla!!! Here we are! We have our collection of books for our poor 1 shelfed library.*
