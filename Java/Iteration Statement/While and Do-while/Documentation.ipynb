But what if she asks you to print 100 numbers or even 10000. Will you sit and write 10000 lines of code? No, it is not possible and that’s where the looping statements come to the rescue.
Now, let us discuss how while and do while loop are implemented in JAVA.


**WHILE LOOP:**
	
while loop is used to execute a set of statements after checking the specified condition. 
  The statements are executed until the condition is true and when the condition becomes false the control jumps to the next line following the code.

 It is also known as entry control loop as the condition is checked at first.






**SYNTAX:**

```
while (condition)
{
	statements to be executed;
	update statement(increment/decrement);
}
```

**FLOW OF WHILE LOOP:**

image.png
 


EXAMPLE 1: 

````python
//JAVA program to print 1 to 5 using while loop
public class Main
{
	public static void main(String[] args)
	{
		int i = 1; // initializing the control variable
		while (i<=5)
		{
			System.out.println(i);
			i++;  // updating the control variable
		}
	}
}

````

Look how only a single print statement is used to display 5 lines instead of using 5 separate print statements. In the same way, we can print 100 or more lines repeatedly by using looping statements.

EXAMPLE 2:

````python
//JAVA program to display the elements of an array using while loop
public class While_loop
{
	public static void main(String[] args)
	{
		int i = 1;
String[] week_days = {"Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"};
		//Using while loop to iterate over an array
		while (i<=7)
		{
			System.out.println("Day"+i+" : "+week_days[i-1]);
			i++;
		}
		System.out.println("Loop terminated");
	}
}

````

Once the loop is executed, the next line following the loop gets executed. Here, after iterating over the week_days array, the next print statement is executed displaying “Loop terminated” in the output.

**DO WHILE LOOP:**
	
  do..while loop is more like the while loop but the specified condition is checked at last i.e after the executing the set of statements.

Since the condition is checked at the end of the loop, it is known as exit control loop and also the body of the loop gets executed at least once even when the condition becomes false.


**SYNTAX:**

```
do 
{
	statements to be executed;
	Update statement(increment/decrement);
} while (condition);
```




**FLOW OF DO WHILE LOOP:**

image.png


EXAMPLE 1:

````python
//JAVA program to print 10 to 15 using do..while loop
public class Main
{
	public static void main(String[] args)
	{
		int i = 10;  //initializing the control variable
		do
		{
			System.out.println(i);
			i++;  // incrementing the control variable 
		} while (i<=15); //condition checking
	}
}

````

EXAMPLE 2:

````python
//JAVA program to illustrate do while loop
public class Dowhile_loop
{
	public static void main(String[] args)
	{
		int i = 26;
		do
		{
			System.out.println("Value of i: "+i);
			i++;
		}while (i<=15);
		System.out.println("Loop terminated");
	}
}

````

Here, the statement is first executed and after when the condition is checked the loop is terminated.



***Note:*** 
You must ensure that the loop will exit after sometime or else you might end up in an infinite loop and program will run forever.

EXAMPLE: 

````python
//JAVA program to illustrate infinite loop
public class Main
{
	public static void main(String[] args)
	{
		int i = 1; //control variable initailization
		while (i<=5)
		{
			System.out.println(i);
		}
	}
}

````

Here, the while loop will run for an infinite time since the variable is not incremented and the condition is always satisfied. So, you must not forget to add update statement.
