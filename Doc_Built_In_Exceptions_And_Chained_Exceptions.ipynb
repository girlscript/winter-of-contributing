##Built-in-exceptions
These exceptions are pre-defined and are present in java libraries. Appropriate exception class is called by compiler automatically when conditions associated with that exception are encountered during run-time or compile-time.

They are of two types:
1.Unchecked exceptions 
2.Checked exceptions

####2.Class cast exception:
 This exception is thrown when we try to downcast parent object to child object or if we try to convert an object of a class to another class's object that have no relationship with each other.
 
```
public class Person {
   //demo-code 
    
}
public class Teacher extends Person{
    //demo-code 
}
public class Singer extends Person{
    //demo-code 
}
public class Main {  
public static void main(String[] args) {  
    Person p=new Person(); //created an object p of type Person
    Teacher t=new Teacher(); //created an object t of type Teacher
    Singer s=new Singer();//created an object s of type singer
    
    p=t; //This will work as teacher is a person.This is called upcasting. 
    t=s; //This is give incompatible type error as even though teacher and singer both extends person class, there is no direct relationship between these two classes.
    s=p; //This will throw an exception as person is not a singer.Here we are trying to typecast parent class object to child class.
  }
}

```
This exception can be solved by following steps:
```
p=s; //Here Singer object has been upcasted to Person. So address of s is now pointed to p.
Singer s1=(Singer)p; //no exception as during runtime it is confirmed that person is a Singer.
```

##Checked Exceptions:
Unlike unchecked exceptions,checked exceptions are thrown during compile time.If a method is throwing this exception it needs to be handled by ways mentioned below otherwise we will get a compilation error.

1.By using throws keyword:

Here we are explicitly throwing ClassNotFoundException by using throw keyword. Here if we do not use throws we will get an compilation error as ClassNotFoundException() being an checked exception needs to be handled by the programmer. 
```
public class MyClass {
    public static void main(String args[]) throws ClassNotFoundException{
      throw new ClassNotFoundException();
    }
}
```
Notice if this was an unchecked exception then there was no need to use throws as java handles unchecked exception on their own.
```
public class MyClass {
    public static void main(String args[]) throws ClassNotFoundException{
      throw new ArithmeticException();
    }
}
```
This compiles successfully because an unchecked exception is thrown.

2. By using try-catch block:

```
public class MyClass {
    public static void main(String args[]){ try{
      throw new ClassNotFoundException();
      //System.out.println("Try block is executed"); //This statement will give error. This is an unreachable statement as the control goes to catch block after exception ClassNotFoundException is thrown.
    }
    
    catch(ClassNotFoundException e){
        System.out.println(e);
        System.out.println("Catch block is executed");
    }
}}
```

##Chained Exception:
Chained exceptions is when an exception is thrown due to occurence of another exception. It is like giving more information to user as to why an excepton is thrown by displaying the cause of that exception. Example-if we obtain an ArithmeticException then it might be because of invalid input from user that is leading to division by 0, so we throw IOException as well.  

There are two methods in throwable class to display the cause of thrown exception.

i)initCause() : It is used to throw the exception that is the cause of our main exception.This method takes cause exception as its parameter. In above example IOException is the cause exception and hence will be passed as a parameter in this method. 

ii)getCause(): It returns the cause that is passed in initCause() method.


Below is another example of Chained exception concept. Here NullPointerException is thrown and its cause that is ClassNotFoundException is thrown by using the methods mentioned above.


```
public class MyClass {
    public static void ChainedException(){
        NullPointerException e=new  NullPointerException();  //Creating object 
        e.initCause(new ClassNotFoundException("caused this exception")); //Initialising cause and passing display statement as argument
        throw e; //throwing NullPointer exception
    }
    public static void main(String args[]) {
       try{
           ChainedException(); //Using try-catch block to catch  NullPointer exception
       } 
       catch(NullPointerException e)
       {
           System.out.println("caught : "+e); //Will display the initial exception that was thrown i.e.NullPointer exception
           System.out.println("cause : "+e.getCause()); //Will display the cause. 
       }
    }
}
```

An easy way to create a chained exception is to directly pass throwable instance while object creating. Eg:
```
import java.io.IOException;

public class ChainedException {

    public static void main(String[] args) {
        try {
            throwException();
        }
        catch (Exception e) {
            System.out.println(e.getCause());
        }
    }

    static void throwException() {
        throw new RuntimeException(new IOException());
    }
}
```
