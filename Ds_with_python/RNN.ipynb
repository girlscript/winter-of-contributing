Introduction to RNNs

A recurrent neural network (RNN) is a type of artificial neural network which recognizes and uses data's sequential characterstics or time series data to detect patterns. (A time series data is a sequence of data points that occur in successive order over a time period like temperature, stock prices etc.) This deep learning algorithm uses training data to learn and develop feedback loops. This is commonly used for problems like language translation, natural language processing (nlp), speech recognition and image captioning . In simple words, RNN is a type of neural  network where  the output from a previous step are fed as input to the current step and it remembers information through time. In cases like when it is required to predict the next word of a sentence, the previous words are required which is where the recurrent nueral network comes into action, unlike in traditional neural networks, where all the inputs and outputs are independent of each other.

How does RNN work?

The artificial neural networks are interconnected and designed with data processing components to function like our human brain. The artificial neurons (network nodes) have the ability to process input and forward output to other nodes in the network. The layers of these neurons are capable of processing information in two directions. RNN consists of a "hidden layer", which remembers the information of sequences. 

                                INPUT ----> HIDDEN LAYER ----> OUTPUT

It uses the same parameters for each input as it performs the same task on all inputs and produces an output, thereby reducing the complexity of increasing parameters. At the same time, it memorizes each previous output, by giving each output as input to the next hidden layer. Hence, they are joined to form a single recurrent layer. 

At the heart of an RNN is a layer made of memory cells. The most popular cell at the moment is the Long Short-Term Memory (LSTM) which maintains a cell state as well as a carry for ensuring that the signal (information in the form of a gradient) is not lost as the sequence is processed. At each time step the LSTM considers the current word, the carry, and the cell state.


TYPES OF RNNs

1.PNG

1) One-to-one:
This is also called Plain Neural networks. It deals with a fixed size of the input to the fixed size of output, where they are independent of previous information/output.

Example: Image classification.

2) One-to-Many:
It deals with a fixed size of information as input that gives a sequence of data as output.

Example: Image Captioning takes the image as input and outputs a sentence of words.

3) Many-to-One:
It takes a sequence of information as input and outputs a fixed size of the output.

Example: sentiment analysis where any sentence is classified as expressing the positive or negative sentiment.

4) Many-to-Many:
It takes a Sequence of information as input and processes the recurrently outputs as a Sequence of data.

Example: Machine Translation, where the RNN reads any sentence in English and then outputs the sentence in French.

5) Bidirectional Many-to-Many:
Synced sequence input and output. Notice that in every case are no pre-specified constraints on the lengths sequences because the recurrent transformation (green) is fixed and can be applied as many times as we like.

Example: Video classification where we wish to label every frame of the video.
