# **Tokenizing**

---

It’s your first step in turning unstructured data into structured data, which is easier to dissect. 
A token is the smallest unit of a sentence. By tokenizing, you can divide the text by word or by phrases. This will allow you to work with smaller pieces of texts that are still fairly coherent and meaningful indeed outside of the environment of the rest of the text. 
 
 Let's us see the types of Tokenizing#

**1. Tokenizing by word-**  Words are like the blocks of natural language. They’re the lowest unit of meaning that still makes sense on its own. Tokenizing your textby word allows you to identify words that come over particularly frequently. For illustration, if you were analyzing a group of job advertisements, also you might find that the word “Developer” comes up frequently. That could suggest high demand for front-end skills, back-end coding but you’d need to look deeper to know more. 

 
**2. Tokenizing by Sentence/Phrases-** When you tokenize, you can dissect how those words relate to one another and see the further correlation.

Write the below command for performing tokenizing:


*from nltk.tokenize import sent_tokenize, word_tokenize*


```python
example_string= """The greatest obstacle to discovery is not ignorance. It is the illusion of knowledge."""
```

```python
#For Sentence Tokenizing

sent_tokenize(example_string)
```

This will give you two sentence as

[ "The greatest obstacle to discovery is not ignorance." 

"It is the illusion of knowledge."]

```python
#For Word Tokenizing
word_tokenize(example_string)
```

This will give all the words of the phrase like this

["The", "greatest", "obstacle", "to", "discovery", "is", "not", "ignorance", ".","It", "is", "the", "illusion", "of", "knowledge", "."]

#**Parts of Speech** 

---

Part of speech is a grammatical term that deals with the places words play when you use them together in sentences. POS trailing, is the task of labelling the words in your text according to their part of speech. 
 
 In English, there are eight parts of speech that are as follows:


 Part of speech
 	Role
   	Examples

**Noun**	

Is a person, place, or thing

eg. mountain, bagel, Poland

**Pronoun**

Replaces a noun

eg. you, she, we


**Adjective**

Shows characteristics

eg. efficient, windy, colorful


**Verb**

Is an action or a state of being	

eg.learn, is, go


**Adverb**

Gives information about a verb, an adjective, or another adverb	

eg. efficiently, always, very


**Preposition**	

Shows connection between a noun or pronoun 	

eg.from, about, at


**Conjunction**	

Connects two other words or phrases	

eg. so, because, and


**Interjection**	
Is an exclamation	

eg. yay, ow, wow
.

# **Lemmentization**

---

A lemma is a unit that represents a whole group of words, and that group of words is called a lexeme. Lemmatizing reduces words to their core meaning, but it will give you a complete English word that makes sense on its own instead of just a fragment of a word 
Let's see how to import it in NLTK:

```python
>>> from nltk.stem import WordNetLemmatizer
```

Now create a lemmantizer to use like this-

```python
>>> lemmatizer = WordNetLemmatizer()
```

```python
>>> lemmatizer.lemmatize("blending")
'blend'
```

# **Analyzing Text through NLTK**

---
First we'll specify a web page and then check out what this web page is about.
We'll be using urilib module to crawl the webpage.


```python
import urllib.request
response =  urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')
html = response.read()
print(html)
```

Now we'll be using a popular Python library Beautiful Soup for pulling data out of HTML and XML files through the below code.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip = True)
print(text)
```

Now let's convert this text into tokens as we've seen in the previous examples.

```python
tokens = [t for t in text.split()]
print(tokens)
```

NLTK provides a function FreqDist() which will count the frequency of the words. Also, we will remove stop words (a, at, the, for etc) from our web page as we don't need them to hamper our word frequency count and then plot the graph for most frequently occurring words in the webpage in order to get the clear picture of the context of the web page

```python
from nltk.corpus import stopwords
sr= stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        
        clean_tokens.remove(token)
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print(str(key) + ':' + str(val))
freq.plot(20, cumulative=False)
```
