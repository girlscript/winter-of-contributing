## Multivariate Adaptive Regression Splines(MARS)

### What is Linear Regression:
- Linear Regression is an algorithm in machine learning used to predict value of continous output.
- Used in  cases where we have to predict the value.
- Works only in labelled data.

### Mathematics Behind Linear Regression

- Fundamental Mathematics behind regression is linear algebra.
- It makes use of the function y=mx+c(for univariate) to find the relation between the dependent variable(predited value) and feature set.


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
import pandas as pd
```

```python
df=pd.read_csv('house.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.lmplot(x='area',y='price',data=df)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

## About MARS(Multivariate Apaptive Regression Splines)
- Used for Non-linear regression problems.
- We can extend linear models to capture any non-linear relationship, which is typically done by using hgher order polynomials(e.g x^2)
- Maimizes predictive accuracy compared to linear regression.
