```python
#import necessary modules
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import Birch
```

```python
#make initial clusters of data points in a gaussian plane using make_blobs
dataset, clusters = make_blobs(n_samples = 1000, centers = 5, cluster_std = 0.50, random_state = 0)
dataset.shape
```

```python
#visualize these clusters
plt.scatter(dataset[:, 0], dataset[:, 1], c = clusters)
```

```python
#initialize the birch model using appropriate parameter values
model = Birch(branching_factor = 50, n_clusters = None, threshold = 1.5)
```

```python
#fit dataset onnto this model
model.fit(dataset)
```

```python
#make final predictions using this model and visualize
pred = model.predict(dataset)
```

```python
#you can compare the plotting with the previous one to find out the the color are different in these cases thereby stating that these clustering are done by different means
plt.scatter(dataset[:, 0], dataset[:, 1], c = pred)
```
