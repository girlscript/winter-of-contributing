```python
#import necessary libraries
from numpy.random import normal
from scipy.stats import norm
from matplotlib import pyplot
```

```python
# define the distribution
mu = 50   #mean
sigma = 5 #standard deviation
n = 10
# generate the sample
sample = normal(mu, sigma, n)
print(sample)
```

```python
#define a normal distribution using the norm() SciPy function 
dist = norm(mu, sigma)
print(dist)
```

```python
values = [i for i in range(30,70)]
pprobs = [dist.pdf(value) for value in values]

# plot pdf
pyplot.plot(values, pprobs)
pyplot.show()
```

```python
# plot cdf
cprobs = [dist.cdf(value) for value in values]
pyplot.plot(values, cprobs)
pyplot.show()
```
