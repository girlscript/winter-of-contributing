# Canny Edge Detection

This is an algorithm that works in multiple steps and is used to detect a variety of edges present in images. Python's OpenCV library combines all the different steps in this algorithm into one single function

cv2.Canny(image, edges, threshold1, threshold2, apertureSize = 3, L2gradient = false)

Here,

image - The 8-bit input image

edges - The 8-bit image after the edges have been detected

threshold1 & threshold2 - The minimum and maximum threshold. The edges with intensity gradients above and below these values are discarded.

apertureSize - This is used in the Sobel filter to find the intensity gradient of the input image

L2gradient - This specifies the equation for finding gradient magnitude

```python
#Importing all necessary libraries
import cv2
import matplotlib.pyplot as plt
```

```python
#Reading the image
img = cv2.imread("puffin.jpg")
plt.imshow(img)
```

```python
#Reading the image, converting it to grayscale and blurring slightly
#This improves the edge detection algorithm
img=cv2.imread("puffin.jpg")
imgray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imblur=cv2.GaussianBlur(imgray,(5,5),0)
plt.imshow(imblur)
```

```python
#Canny edge detection for a wide threshold 
c1=cv2.Canny(imblur,10,50) #Only input image and threshold values are mentioned
plt.imshow(c1)
```

```python
#Canny edge detection for a mid-ranged threshold
c2=cv2.Canny(imblur,100,180)
plt.imshow(c2)
```

```python
#Canny edge detection for a narrow threshold
c3=cv2.Canny(imblur,240,250)
plt.imshow(c3)
```
