```python
import cv2
import matplotlib.pyplot as plt
```

ðŸŒŸ Read an image from path

cv2.imread(path,flag)

Parameter:

â€¢ path: Location of image to be read as a string.
    
â€¢ flag(optional): flag value, that defines the format image to be read. Default value is 1(CV2.IMREAD_COLOR).It value varies from 0, 1, -1.

```python
image = cv2.imread('doggy.jpg')
plt.imshow(image[:,:,::-1])  
plt.show()
```

1. Blurring an Image

ðŸŒŸ Gaussian Blur:

cv2.GaussianBlur(image,ksize,sigmaX,sigmaY,borderType)

Parameter:

â€¢ image: source image read
    
â€¢ ksize: kernel size as tuple
    
â€¢ sigmaX(optional): kernel standard deviation along x-axis
    
â€¢ sigmaY(optional): kernel standard deviation along y-axis
    
â€¢ borderType: specifies image boundaries
    
    
ðŸŒŸ Median Blur:

cv2.medianBlur(image, ksize)

Parameter:

â€¢ image : source image read
    
â€¢ ksize: kernel size as integer
    
    
ðŸŒŸ Bilateral Blur:

cv2.bilateralFilter(image, d, sigmaColor, sigmaSpace, borderType)

Parameter:

â€¢ image: source image read
    
â€¢ d: diameter of pixel neighbourhood as integer
    
â€¢ sigmaColor: filter sigma in the color space as integer
    
â€¢ sigmaSpace: filter sigma in the coordinate space.
    
â€¢ borderType: specifies image boundaries

```python
image = cv2.imread('doggy.jpg')
Gaussian_image = cv2.GaussianBlur(image, (11,9), 0)
plt.imshow(Gaussian_image[:,:,::-1])  
plt.show()
```

```python
median_image = cv2.medianBlur(image, 7)
plt.imshow(median_image[:,:,::-1])  
plt.show()
```

```python
bilateral_image = cv2.bilateralFilter(image, 9, 75, 75)
plt.imshow(bilateral_image[:,:,::-1])  
plt.show()
```

ðŸŒŸ Grayscaling of images

cv2.cvtColor(image,colorcode)

Parameter:

â€¢ image: source image read
    
â€¢ colorcode: color space conversion code. for grayscale colorcode is cv2.COLOR_BGR2GRAY

```python
image = cv2.imread('doggy.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY )
plt.imshow(gray_image,cmap='gray')  
plt.show()
```

```python
gray1_image = cv2.imread('doggy.jpg',0)
plt.imshow(gray1_image,cmap='gray')  
plt.show()
```

ðŸŒŸ Image Resizing

cv2.resize(image,(width,height),fx,fy,interpolation)

Parameter:

â€¢ image: source image read
    
â€¢ (width,height) : dimension as tuple
    
â€¢ fx(optional) : scaling along horizontal axis.
    
â€¢ fy(optional) : scaling along vertical axis.
    
â€¢ interpolation(optional) : flag value that defines the any following method -INTER_NEAREST, INTER_LINEAR, INTER_AREA, INTER_CUBIC.

```python
image = cv2.imread('doggy.jpg')
resized_image = cv2.resize(image,(500,800))
plt.imshow(resized_image[:,:,::-1])  
plt.show()
```
