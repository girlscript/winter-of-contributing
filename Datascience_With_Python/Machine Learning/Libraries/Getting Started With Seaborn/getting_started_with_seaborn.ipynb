# Getting started with Seaborn

# What is Seaborn?

Seaborn is an open source liabrary of python. It is used for data visualization and explratory data analysis. It includes different types of charts which helps to get the relationship between variables in a dataset and also provides plots to get insights from data. Seaborn is data visualization library based on matplotlib. Visualization is the central part of Seaborn which helps in exploration and understanding of data. Seaborn provides beautiful default styles and colour palettes to make statistical plots more attractive.

Seaborn helps you to generate different plots based on the data you are using. There are plenty of excellent Python visualization libraries available, including the built-in matplotlib. But seaborn stands out for me. It combines aesthetic appeal seamlessly with technical insights.

# Why you should use Seaborn versus matplotlib?

Seaborn makes our charts and plots look engaging and enables some of the common data visualization needs like mapping color to a variable. It makes the data visualization and exploration easy to conquer.
Seaborn comes with a large number of high-level interfaces and customized themes that matplotlib lacks as it’s not easy to figure out the settings that make plots attractive.
Matplotlib functions don’t work well with dataframes, whereas seaborn does.

### Importing necessary libraries using following code:

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':(10,6)})
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

Reading data from 'Iris' dataset using pandas read_csv method.

```python
df = pd.read_csv(r"iris.csv")
df.head()
```

Reading data from another dataset i.e. 'tips' using pandas read_csv method.

```python
df1 = sns.load_dataset('tips')
df1.head()
```

## Different types of plots in seaborn

### Bar Plot

Bar plot shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.
It is also important to keep in mind that a bar plot shows only the mean value, but in many cases it may be more informative to show the distribution of values at each level of the categorical variables.

```python
sns.barplot(x=df["Species"], y=df['SepalLengthCm'],hue=df["Species"])
plt.show()
```

As shown in the above bar chart that there are 3 different types of speices which are plotted on the x-axes and column SepalLengthCm on y-axes.

Hue property is used which helps anyone to understand the chart easily and can identify the results with the help of the chart.

### SwarmPlot

Swarm plot function always treats one of the variables as categorical and draws data at ordinal positions (0, 1, … n) on the relevant axis, even when the data has a numeric or date type.

```python
sns.swarmplot(x='Species', y='SepalLengthCm',data=df)
plt.show()
```

### CountPlot

When you use sns. countplot , Seaborn literally counts the number of observations per category for a categorical variable, and displays the results as a bar chart. Essentially, the Seaborn countplot() is a way to create a type of bar chart in Python.

```python
sns.countplot(x="Species",data=df,palette="rainbow")
plt.show()
```

Looking at the plot we can say that the number of species of different category have total number of count as 50 in the dataset. As it only returns the count based on a categorical column, we need to specify only the x parameter having catergorial column.

### BoxPlot

Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used to detect the outlier in the data set.

```python
sns.boxplot(x="day",y="total_bill",data=df1,hue="smoker")
plt.show()
```

x takes the categorical column and y is a numerical column. Hence we can see the total bill spent each day.” hue” parameter is used to further add a categorical separation. By looking at the plot we can say that the people who do not smoke had a higher bill on Friday as compared to the people who smoked.

### ViolinPlot

It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution.

```python
sns.violinplot(x='day', y='total_bill', data=df1, hue='sex', split=True)
plt.show()
```

hue is used to separate the data further using the sex category
setting split=True will draw half of a violin for each level. This can make it easier to directly compare the distributions.

### Strip Plot

A strip plot is a scatter plot where one of the variables is categorical. The strip plot is an alternative to a histogram or a density plot. It is typically used for small data sets.

```python
sns.stripplot(x='day', y='total_bill', data=df1, hue='smoker', jitter=True, split=True)
plt.show()
```

One problem with strip plot is that you can’t really tell which points are stacked on top of each other and hence we use the jitter parameter to add some random noise.

jitter parameter is used to add an amount of jitter (only along the categorical axis) which can be useful when you have many points and they overlap so that it is easier to see the distribution.

hue is used to provide an additional categorical separation.

### Point Plot

This plot points out the estimate value and confidence interval. Pointplot connects data from the same hue category. This helps in identifying how the relationship is changing in a particular hue category.

Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables.

```python
sns.pointplot(x='day', y='total_bill', data=df1, hue='smoker')
plt.show()
```

As we can see from above figure that point is created in between and they are connected for the categorial column depending upon the value of the hue.

This plot is effective in finding the mean and connect them according to the category column.

### Dist Plot

Dist plot gives us the histogram of the selected continuous variable.

It is an example of a univariate analysis.

We can change the number of bins i.e. number of vertical bars in a histogram.

```python
sns.distplot(x=df1["total_bill"],bins=8)
plt.xlabel("Total bill")
plt.show()
```

We can see in above chart that it creates number of bins that we specify during plot.

It draws histograms and fits a Kernel Density Estimate i.e. kde whose value if by default True.

On x axis there is total_bill column from dataset and on y axis it is density which is taken by default.

### Joint Plot

Joint plot is the combination of the distplot of two variables.

It is an example of bivariate analysis.

We additionally obtain a scatter plot between the variable to reflecting their linear relationship. We can customize the scatter plot into a hexagonal plot, where, more the color intensity, the more will be the number of observations.

```python
sns.jointplot(x=df1["day"], y=df1["total_bill"])
plt.show()
```

We can see the linear relation between day and total_bill.

We can also use different values of kind that are kind="hex" or kind="reg" where hex provides hexagonal plot and reg provides a regression line on the graph.

### Pair Plot

It takes all the numerical attributes of the data and plot pairwise scatter plot for two different variables and histograms from the same variables.

Plot pairwise relationships in a dataset. By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column.

```python
sns.pairplot(data=df1)
plt.show()
```

As shown in above plot that histograms are created for the variables which are same on both the axes.

Pair plot is used to understand the best set of features to explain a relationship between two variables or to form the most separated clusters.

### Rug Plot

It draws a dash mark instead of a uniform distribution as in distplot.

It is an example of a univariate analysis.

```python
sns.kdeplot(x="total_bill", data=df1)
sns.rugplot(x="total_bill", data=df1)
plt.show()
```

As seen we have used rug plot for the total_bill column and also used kde plot together where x axis contains total_bill and y axis contains density by default.

We can see clearly that the density is higher for the value of total_bill i.e. 15.

### Heat Map

This plot comes under Matrix Plot which is 2-dimensional matrix.

For heat map we will use correlation method which will correlate numeric data from dataset and will plot it on the chart which will help to read the values easily.

Heat map also gives us to use color for interpretation to become easy.

```python
sns.heatmap(df1.corr(), annot=True, cmap='viridis')
plt.show()
```

A heatmap contains values representing various shades of the same colour for each value to be plotted. Usually the darker shades of the chart represent higher values than the lighter shade.

For the same variables it shows the value as 1.

It will display completely different color for very different value.

The primary purpose of Heat Maps is to better visualize the volume of locations/events within a dataset and assist in directing viewers towards areas on data visualizations that matter most.
