## Special functions

A large number of mathematical special functions are important for many computional physics problems. SciPy provides implementations of a very extensive set of special functions.

It includes Cubic Root, Exponential, Log sum Exponential, Lambert, Permutation and Combinations, Gamma, Bessel etc

```python
#Cubic Function:

from scipy.special import cbrt #Find cubic root of 49 & 64
cb = cbrt([49, 64] ) #print value of cb
print(cb)
```

```python
#Exponential function

from scipy.special import exp10 #define exp10 function and pass value in its
exp = exp10([10,100])
print(exp)

```

```python
#Permuations and Combinations

#Permuation

from scipy.special import perm 
per = perm(6, 3, exact = True)
print(per)

#Combination 
from scipy.special import comb
com = comb(6, 3, exact = False, repetition=True)
print(com)
```

**#Linear equation systems on the matrix form**

Ax=b 

where  A  is a matrix and  x,b  are vectors can be solve

```python
import numpy as np  
from scipy import linalg  
# the input matrix
#         x -3y +z = 11
#         x -2y + 6z = 9
#         -5x + 18y + 7z = 80  
# Creating input array  
a = np.array([[1, -3, 1], [1, -2, 6], [-5, 18, 7]])  
# Solution Array  
b = np.array([[11], [9], [80]])  
# Solve the linear algebra  
x = linalg.solve(a, b)  
# Print results  
print(x)  
 
```

***Fourier Transform***

Fourier transform is a complex technique which is used in converting spatial domain data into frequency domain data. 
FFT or Fast Fourier Transform is a algorithm of DFT. SciPy provides functions for accessing the classic FFTPACK library. It is applied in ndimensional array.

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np 

#Lets periodic function example of cos(20 Ã— 2Ï€t)

fre  = 5 #Frequency in terms of Hertz
fre_samp = 50 #Sample rate
t = np.linspace(0, 2, 2 * fre_samp, endpoint = False )
a = np.cos(fre  * 2 * np.pi * t)
figure, axis = plt.subplots()
axis.plot(t, a)
axis.set_xlabel ('Time (s)')
axis.set_ylabel ('Amplitude')
plt.show()


#Visualizing magnitude of the signal

from scipy import fftpack

main= fftpack.fft(a)
frequency = fftpack.fftfreq(len(a)) * fre_samp
figure, axis = plt.subplots()

axis.stem(frequency, np.abs(main))
axis.set_xlabel('Frequency in Hz')
axis.set_ylabel('Frequency Spectrum Magnitude')
axis.set_xlim(-fre_samp / 2, fre_samp/ 2)
axis.set_ylim(-5, 110)
plt.show()
```

***Optimization***
> Optimization (finding minima or maxima of a function) is useful algorithm for minimization of curve fitting, multidimensional or scalar and root fitting.

To use the optimization module in scipy first include the optimize module:

```python
#Least Square minimization

#It finds a local minimum of any cost function

from scipy.optimize import least_squares  
import numpy as np  
input = np.array([3, 3])  
def rosenbrock(x):  
   return np.array([10 * (x[1] - x[0]**3), (1 - x[0])])  
res = least_squares(rosenbrock, input)  
print(res)  
```

```python
#Curve fitting Optimization

import numpy as np   
from scipy.optimize import curve_fit   
from matplotlib import pyplot as plt   
x = np.linspace(0,20, num = 50) #for creating numeric sequences.

# The coefficients are much bigger.   
y = 10.35 * np.cos(5.330 * x) + np.random.normal(size = 50)   
def test(x, a, b):   
    return a * np.cos(b * x)   
param, param_cov = curve_fit(test, x, y)   
print("Cos funcion coefficients:")   
print(param)   
print("Covariance of coefficients:")   
print(param_cov)   
ans = (param[0]*(np.cos(param[1]*x)))   
plt.plot(x, y, 'o', color ='red', label ="raw data")   
plt.plot(x, ans, '--', color ='green', label ="optimized data")   
plt.legend()   
plt.show()  
```

**Interpolation****
> The estimation of intermediate value between the precise data points is called as interpolation". Interpolation can be used in scipy as a interp1d function. This function returns an interpolated (Y) value while given arrays in the range of X.

Types of interpolation:
* Linear
* Nearest
* Zero
* S-linear
* Quadratic
* Cubic

```python
#One Dimensional Interpolation
import numpy as np
from scipy.interpolate import interp1d
import pylab

A, nu, k = 10, 4, 2

def f(x, A, nu, k):
    return A * np.exp(k*x) * np.sin(2*np.pi * nu * x) #finding intermediate points of this function

xmax, nx = 0.5, 8
x = np.linspace(0, xmax, nx) #Creating numercial array
y = f(x, A, nu, k)

f_nearest = interp1d(x, y, kind='nearest') 
f_linear  = interp1d(x, y)
f_cubic   = interp1d(x, y, kind='cubic')

x2 = np.linspace(0, xmax, 100)
pylab.plot(x, y, 'o', label='data points')
pylab.plot(x2, f(x2, A, nu, k),  '-', label='exact')
pylab.plot(x2, f_nearest(x2),':', label='nearest')
pylab.plot(x2, f_linear(x2),'-.', label='linear')
pylab.plot(x2, f_cubic(x2),'--', label='cubic')
pylab.legend()
pylab.show()
```

**Statistics**
> The scipy.stats module contains a large number of statistical distributions, statistical functions and tests.

```python
#Descriptive Stat

import scipy as sp  
import numpy as np  
from scipy.stats import norm  
number_of_data = 150
random_data_set = sp.randn(number_of_data) 
print("Mean of this array \n")
print(random_data_set.mean()) #Computes geometric mean  
print("\n Median of this array \n")
print(sp.median(random_data_set))  # Computes geometric median
print("\n Minimum and Maximum of this array \n")
min_max = np.array([random_data_set.min(),random_data_set.max()])  #minimum and maximum value 
print(min_max)
```

***Sparse Matrix Problem***

> The sparse matrix is used for storing large sparse matrices, and it  provide the functionality to perform complex matrix computations.It allows storetos to store only non-zero values considering other values as zeroes.

Ways to represent sparse matrix; 

* Block Sparse Row matrix (BSR)
* Coordinate list matrix (COO)
* Compressed Sparse Column matrix (CSC)
* Compressed Sparse Row matrix (CSR)
* Sparse matrix with Diagonal storage (DIA)
* Dictionary Of Keys based sparse matrix (DOK)
* Row-based linked list sparse matrix (LIL)

```python
import numpy as np  
from scipy.sparse import random  
np.random.seed(40)  
# Generate a random binary sparse matrix  
matrix = random(10,10, format='csr', density=0.25)  
# Substitute all non zero values with index number  
matrix.data[:] = np.arange(1, matrix.data.shape[0]+1)  
# We can access and modify these arrays:  
print(matrix.toarray())  
print("\n The data is ",matrix.data)  

print("\n Cell which consits the non-zero values",matrix.indptr)  
```
