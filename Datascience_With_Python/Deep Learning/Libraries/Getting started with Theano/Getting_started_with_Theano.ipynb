# **GETTING STARTED WITH THEANO**

# **WHAT IS THEANO ?**

Theano is a Python library developed in **2007** by **Yoshua Bengio** and the **reasearchers** at **Université de Montréal**. It was named after a *female Greek Philosopher*, ***Theano of Crotone***. It is mainly used in deep learning tasks and is also called as the "grandfather" of deep learning libraries .

Although Theano is not in much use now, the other libraries built on top of Theano such as Keras, Lasagne, and Blocks are still being used.
Theano is considered to be the industry standard for Deep Learning research and development.

In simple terms Theano can be refered to as an *hybrid* between **NUMPY** and **SYMPY**.

# **WORKS SUPPORTED BY THEANO**

1.   Expression types of Theano covers most of the functionalities of NumPy and SciPy.

2.   Theano supports N-dimensional array of integers (both signed and unsigned) and floating point (both of single precision and double precision) values.

3.   Theano supports arbitary python objects.

4.   Theano supports sparse vectors and matrices via SciPy’s sparse module. 

5.   Theano also supports specialized expression types such as the logistic sigmoid, the softmax function, and multi-class hinge loss for machine learning research.

# **EXAMPLE PROGRAMS USING THEANO**

# **STEPS for a Theano application:**
* Define symbolic expressions
* Compile a function that can compute numeric values using those expressions
* Execute that function on data

***Example 1 :***
```
import theano
from theano import tensor

x = tensor.dscalar()
y = tensor.dscalar()

z = x + y
f = theano.function([x,y], z)
print(f(1.5, 2.5))
```

When we run this the output will be as follows:
```
4.0
```

***Example 2 :***

```
import theano
import numpy
import theano.tensor as T
from theano import function

x = T.dmatrix('x')
y = T.dmatrix('y')
z = x + y
f = function([x, y], z)
 
print(f([[10, 20], [20, 30]], [[40, 50], [30, 40]]))
```
When we run this the output will be as follows:
```
array([[ 30.,  90.],
       [ 50.,  70.]])
```


***Example 3 :***

**Calculating *multiple* results**

```
#import the libraries
import theano
from theano import tensor

# declare variables
x, y = tensor.dmatrices('x', 'y')

# simple expressions for each operation
diff = x - y

abs_diff = abs(diff)
diff_squared = diff**2

# the function operation
f = theano.function([x, y], [diff, abs_diff, diff_squared])

# call the function and store the result in a variable
result= f([[1, 1], [1, 1]], [[0, 1], [2, 3]])

# printing the output
print('Difference: ')
print(result[0])

print('Absolute Difference: ')
print(result[1])

print('Squared Difference: ')
print(result[2])
```

When we run this the output will be as follows:
```
Difference:
[[ 1.  0.]
 [-1. -2.]]
Absolute Difference:
[[ 1.  0.]
 [ 1.  2.]]
Squared Difference:
[[ 1.  0.]
 [ 1.  4.]]
```

>**UNDERSTANDING THE  *function*   OPERATION USED:**


>Theano functions are basically used for interacting with the symbolic graph. They are used for passing input into the network and collecting the output.
```

```
> The first parameter is used to define the list of input variables, and the second parameter is used to define the list of output variables.
