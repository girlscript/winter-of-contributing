#Convert an image into pencil sketch using OpenCV
In this task we will follow five important steps given below to convert an image into pencil sketch using OpenCV.

###OpenCV:
OpenCV (Open Source Computer Vision Library) is an open source computer vision and machine learning software library. OpenCV was built to provide a common infrastructure for computer vision applications and to accelerate the use of machine perception in the commercial products. 

#### Importing OpenCV

```python
import cv2  #OpenCV
from google.colab.patches import cv2_imshow
```

#### Read and display the image

```python
img = cv2.imread('pic1.jpg')   #reading an image
#Image source: https://pixabay.com/photos/girl-portrait-hairstyle-redhead-919048/
```

```python
cv2_imshow(img)  #showing an image
```

#### STEP 1: Convert image into gray using cvtColor function of OpenCV
Gray image: converting an image from other color spaces e.g. RGB, CMYK, HSV, etc. to shades of gray

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  
cv2_imshow(gray_img)
```

#### STEP 2: Invert gray image using bitwise_not function of OpenCV
A bitwise NOT inverts the "on" and "off" pixels in an image.

```python
invert_img = cv2.bitwise_not(gray_img)
cv2_imshow(invert_img)
```

#### STEP 3: Blur the image using GaussianBlur function
The function convolves the source image with the specified Gaussian filter.<br>
The Gaussian filter is used to remove Gaussian noise

```python
smoothing_img = cv2.GaussianBlur(invert_img, (21,21), sigmaX=0, sigmaY=0)
                          #(21,21): Gaussian filter size (it should be a pair of odd numbers)
                          #sigmaX: Gaussian filter standard deviation in X direction
                          #sigmaY: Gaussian filter standard deviation in X direction 
cv2_imshow(smoothing_img)
```

#### STEP 4: Invert the blurred image

```python
invert_blur_img = cv2.bitwise_not(smoothing_img)
cv2_imshow(invert_blur_img)
```

#### STEP 5: Obtain the final pencil sketch image using divide function between gray image and inverted blurred image
divide(): Performs per-element division of two arrays or a scalar by an array

```python
sketch_img = cv2.divide(gray_img, invert_blur_img, scale=256.0)
```

```python
print('Original Image')
cv2_imshow(img)
print('\n\nSketch')
cv2_imshow(sketch_img)
```
