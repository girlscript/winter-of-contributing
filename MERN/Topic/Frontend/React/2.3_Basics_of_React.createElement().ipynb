# Basics of React.createElement()


### Introduction

HTML is a basic set of instructions which specifies the skeleton of website. When HTML elements are loaded in browser, they are called DOM(Document Object Model) elements. These DOM elements are finally rendered in user interface of a website.

```python
from IPython.core.display import display_html
display_html('<h1 class=”hero” id=”hero_heading”> </h1>')
```

 The above HTML is equivalent to 

 let h1 = document.createElement(“h1”);
h1.setAttribute(“class”,”hero”);
h1.setAttribute(“id”,”hero_heading”);

 in JavaScript



However, managing DOM elements using pure JavaScript is too tedious and time consuming.

Here comes React (a JavaScript library) which frees us from this tedious task of interacting with DOM. React uses the concept of virtual DOM, made up of React elements. These look similar to HTML elements but are JavaScript objects. This can be done in two ways:
1.using React.createElement()
2.using JSX


## React.createElement()

### Why is it used?

It is used to create a new React element so that it could be included in the ReactDOM and then rendered. 

eg: React.createElement(“p”,{className:”hero”,id:”hero_para”},”Lorem ipsum”)

What this syntax will do is to create a paragraph element with class “hero” ,id “hero_para” and “Lorem ipsum” as innerHTML or its child.

### When to use it?

Using React without JSX is especially helpful when we don’t want to set up compilation in our build environment.
