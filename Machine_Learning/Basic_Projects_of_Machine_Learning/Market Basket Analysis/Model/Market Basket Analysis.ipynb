```python
basket=pd.get_dummies(data)

basket
```




**1. Support**: percentage of transaction that contain all of the items in an itemsets. 
Higher the support more frequently the itemset occurs.

**Example** Let total number of orders be 20 and {bread,butter} occurs 5 times.

support=  5/20=0.25

```python
#filter items with support more than 0.3%

frequent_sets_1=apriori(basket,min_support=0.003,use_colnames=True)
frequent_sets_1
```

```python
#filter items with support more than 4%

frequent_sets_2=apriori(basket,min_support=0.04,use_colnames=True)
frequent_sets_2
```

# Applying Association Rule

**1. Confidence**: Suppose there are two items bread and butter.Confidence calcutes the number of transactions involving the items bread and butter. In other words, confidence measures the number of times butter is purchased, given that bread is purchased.

{bread=>butter}=frequency(bread,butter)/freq(bread)

**2.Lift**: (x=>y) It is the likelihood of getting item y purchased when item x is sold.

1. If Lift(x=>y)=1 means there is no correlation.
2. If Lift(x=>y)>1 means there is positive correlation within itemset.
3. If Lift(x=>y)<1 means there is negative correlation within itemset.

**3.Conviction**: conv(x => y) = 
1.Conv(x=>y)=1 means that x has no correlation with y

2.Greater the conviction higher the interest in the rule 

**4.Antecedents & Consequent**: An antecedent is an item found within data and consequent is an item found with in combination with antecedent. 

```python
#list of topn itemsets with strong association rule

rules=association_rules(frequent_sets_1,metric="lift",min_threshold=0.1)
rules.head(25)
```

```python
#Listing the itemsets having confidence more than 70% and lift greater than 6
rules[(rules['lift']>6)&(rules['confidence']>0.7)]
```
