### What are Variables?
- A Variable is any quantity that can be measured or counted.
- In terms of data science, a variable can be date, time, quantity,etc.

### Why is it important to know about variables?
- Knowing what type of variables helps us to identify  the operations we can perform if there are missing variables.
- Knowing type of variable helps us identify if the data is faulty.

### Types of Variables:
#### 1.  Numerical
#### 2. Categorical

## Numerical variable
- These types of data types are measurable and are expressed in numbers.

Numerical data types are further classified into:
#### 1. Discrete 
- These types of variables are precise and consist of assigning observations to one of a number of categories in terms of counts or proportions
- Example: number of students.

#### 2. Continous
- These types of variables can take on an uncountable set of values within an interval.
- Example: Age, as it is generally rounded off.

## Categorical variable
- These types of variables are not a single quantity, it can be mean a string of text or date.
- Examples: High, low.

 Categorical variable type is further classified into:
#### 1. Ordinal
- An ordinal variable is a variable whose values are defined by an order relation between the different categories.
- For example "Good", "Very Good".

#### 2. Binary
- A binary variable is a variable which has only 2 values
- Example: (Yes,No),(True,False)

#### 3. Nominal
- A nominal variable describes a label or a category without any order
- Example: Gender, Name

## Range
- Range is the difference of highest and lowest value.
- It helps us to identify how spread out the data is.

#### For example:

```python
#Importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading dataset
df=sns.load_dataset("flights")
df.head() #Checking first 5 values
```

```python
range=df["passengers"].max()-df["passengers"].min()
print(f'Range of passengers is {range}')
```

## Population Distribution
- Population distribution is the study of distributions of the data.
- It includes all of the elements from a set of data
- This method is not applicable for a very large data.
- It helps us to understand the variability of the data.

#### For example:

```python
sns.displot(df,x='passengers')
```

```python
#calculating 
mean=df["passengers"].mean()
print(f'Mean of passengers is {mean}')
```

## Sample Distribution
- If the data is very large we take subset of the data to understand the whole data which is sample distribution.
- It represents the entire data that we want to study.
