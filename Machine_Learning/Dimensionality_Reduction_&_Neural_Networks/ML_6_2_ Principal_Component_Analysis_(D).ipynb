# Principal Component Analysis


## Introduction


It is an unsupervised machine learning algorithm used primarily for dimensionality reduction in machine learning. It is a statistical process that converts the set of correlated features into a set of linearly uncorrelated features with the help of orthogonal transformation.These new features that have undergone the process of transforamtion are called Principal Components.

It is also used for EDA and predictive modeling. It is used to draw strong patterns from dataset by reducing the variance.PCA tries to fit a high dimensional dataset into a lower dimensional surface.

### Real World Applications

Some real-world applications of PCA include image processing, movie recommendation systems, optimizing the power allocation in various communication channels among others.

### Understanding PCA better

The PCA algorithm is based on some mathematical concepts such as:

###### Variance and Covariance
###### Eigenvalues and Eigen factors

##### Common Terms 

Having an idea about these can help us in having a better idea regarding PCA


###### Dimensionality:
It is the number of features or present in the  dataset. In other words, it is the number of columns present in the dataset.

###### Correlation: 
It signifies that how strongly two variables are related to each other.If one changes , the other variable also gets changed. The correlation value ranges from -1 to +1. Here, -1 indicates thats variables are inversely proportional to each other, and +1 indicates that variables are directly proportional to each other.
###### Orthogonal: 
It defines that variables are not correlated to each other, and hence the correlation between the pair of variables is represented by zero.
###### Eigenvectors: 
Considering a square matrix M, and a non-zero vector v.v is called the eigenvector if Av is the scalar multiple of v.
###### Covariance Matrix: 
A matrix containing the covariance (relation) between a pair of variables is called the Covariance Matrix.

###### NOTE:
The no of PCs are either equal to or less than the no of features originally present in the dataset.


##### Properties of Principal Components

###### The principal component must be the linear combination of the original features of dataset.
###### The components are orthogonal, i.e. correlation between a pair of variables is zero.
###### The importance of each component decreases when going from 1 to n, it means the 1st PC has the most importance, and nth PC will have the least importance.

### Steps to perform PCA

###### Getting the dataset
###### Representing data into a structure
###### Standardizing the data
###### Calculating the Covariance of Z
###### Calculating the Eigen Values and Eigen Vectors
###### Sorting the Eigen Vectors
###### Calculating the new features Or Principal Components
###### Remove less or unimportant features from the new dataset
