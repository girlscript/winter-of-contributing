#                                                     Introduction


Supervised learning can be categorized into two algorithms namely classification and regression. Supervised Learning refers to the learning in which the model learns with the help of labeled data ie. the data used is labeled (the labeled data can be a result of human intervention or can be a product of unsupervised learning model) 

Both are primarily used for predictive analysis in machine learning.

The major difference between the above two algorithms are that regression is used for prediction of continous values like price , salary , age etc while classification is used to predict discrete values like malignant or not , True or False ,Yes or No etc



# Classification

Classification is the process of creating a function that helps in dividing the dataset into classes.In this process the model or code is trained using data called training data and based on this training it is able to create classes.To better understand classification we can consider the example of spam detection

##### Example

The model is trained on the basis of millions of emails on different parameters(assessing criteria) and whenever it receives a new email , it identifies whether the email is spam or not. If the email is spam, then it is moved to the Spam folder.

###### Random Forest Classification

## Regression

While implementing Regression we try to find correlation between the independent and the dependent variables. The dependent variables are the variables whose values change upon changing the value of other variables while the independent variables are the variables whose values are unrelated to other variables and so they don't undergo a change when we change other variables in the equation. We can consider the example of house prediction to understand the regression algorithm.

##### Example 

Suppose we want to predict the price of the house then we need to look upon the data that contains information about the size of the house, the locality, cost per square meter ,etc. This data will also contain some house prices and the relevant data to match so that the model after training on the data can predict the house prices for new unseen data.The house price here is the dependent variable whereas the parameters like size of house and cost per square meter are independent variables.
