# **Why Python Programming**

> Python was created in 1991 by a Dutch Programmer, Guido Van Rossum.

> Python is an Object Oriented Programming Language
* Which means we create objects and classes, which can be reused to implement data and output according to the users requirements.
* But, Python can also be treated in a procedural or functional way.

> Python is an Open source programming language
* It is developed under an OSI(Open Systems Interconnection) which is an approved open source license
* It is free to access and download for everyone.
* It is free to use and distribute even for commercial purpose.

> Python is a High-level language
* It is easily understandable by humans.

> Python is Easy to learn.
* Its syntax is similar to English Language.
* Semicolon is not used at the end, this reduces the probability of warnings and errors.
* It has many Frameworks, tools and libraries
* It takes hardly 1-2 months to well understand the fundamentals of Python.

> Python is an Interpreted language.
* It uses interpreter instead of compiler. Languages like C, C++ uses the compiler.
* Interpreter translates one statement of the program at a time by converting byte code into machine code.
* It is faster than compiler.
* Error detection is easy in interpreter it does not generate an intermediary code.

> Python is Portable.
* This means any source can implement the code written in Python.
* Python works on different platforms, i.e.Windows, Linux, Mac.
* Python can run without installing it from any open source or by USB.

> Python’s Execution speed is faster.
* As it uses interpreter its execution speed is faster from other languages which uses compiler.

> Python has many Frameworks, tools and libraries available 
* This makes the coding in Python easier
* Django and Flask for developing web applications, 
* TensorFlow for deep learning, 
* Matplotlib for plotting charts and graphs, 
* Scipy for mathematics applications, 
* BeautifulSoup for Web Scraping and 
* Pandas for data analysis.
* Python is the best for automation and software testing because it just needs few lines of code to display required results.

> Python helps to code in fewer(minimal) lines
* For example: Hello World code can be completed in just one line as shown below
* print(“Hello World”)

> Python is Versatile.
* It can be used for Web Development, Machine Learning, Software Development, Artificial Intelligence, Mathematics, GUI Application Development, Game Development.
* Where, Big Data, Machine Learning, AI, Cloud computing are the ongoing trends in the technological world.
* It is a good platform for data science.

> Python is highly productive if compared to other OOP languages like Java and C++.
* It has unique feature of handling datatypes on its own.
* Python is much faster in terms of development.
* Python is reliable more than any other programming languages.

> Python has a large community.
* That community supports and develops the ecosystem by adding their contributions and libraries. 
* Easy Documentation and developer support is available for all levels beginner to expert to learn and adapt this programming language.
* Many of the data processing workloads in the organization are powered by python programming language.

> Python is High in demand in the development market. 
* Therefore, it has wide range of opportunities opened for python developers with high salary packages.
* Python is a valueable skill-set to have for any developer or programmer.
* Google is the supporter of Python. It adopted python programming in 2006.
* Many big organizations like Facebook and Amazon Web Services supports Python.
* Python is used by popular websites like Youtube, Reddit, Dropbox, Instagram etc.
