# BITWISE OPERATORS

Python comes with a few different kinds of operators, such as the arithmetic, logical, and comparison operators. You can think of them as functions that take advantage of a more compact prefix and infix syntax.

There are following Bitwise operators supported by Python language.

|OPERATOR|DESCRIPTION|SYNTAX|
|------|------|------|
|& |	Bitwise AND |X & Y  | 
| \| | 	Bitwise OR | X \| Y | 
|~|	Bitwise NOT  | ~X  | 
|^ |	Bitwise XOR  |X ^ Y  | 
|>> |	Bitwise RIGHT SHIFT  |Y>>  | 
|<<| 	Bitwise LEFT SHIF  | Y<< | 


Lets try to understand Bitwise operator in details
Bitwise AND operator: Returns 1 if both the bits are 1 else 0.
Bitwise or operator: Returns 1 if either of the bit is 1 else 0
Bitwise not operator: Returns oneâ€™s complement of the number.
Bitwise xor operator: Returns 1 if one of the bits is 1 and the other is 0 else returns false


```bash
x = 24
y = 22
 
# Print bitwise AND operation
print("x & y =", x & y)
 
# Print bitwise OR operation
print("x | y =", x | y)
 
# Print bitwise NOT operation
print("~x =", ~x)
 
# print bitwise XOR operation
print("x ^ y =", x ^ y)


```

## Bitwise Shift Operators

These operators are used to shift the bits of a number left or right thereby multiplying or dividing the number by two respectively. They can be used when we have to multiply or divide a number by two

Bitwise Left shiftt:The bitwise left shift operator (<<) moves the bits of its first operand to the left by the number of places specified in its second operand. It also takes care of inserting enough zero bits to fill the gap that arises on the right edge of the new bit pattern

Bitwise Right Shift:The bitwise right shift operator (>>) is analogous to the left one, but instead of moving bits to the left, it pushes them to the right by the specified number of places. The rightmost bits always get dropped

```bash
x = 4
y = -5

# print bitwise right shift operator
print("a >> 1 =", x >> 1)
print("b >> 1 =", y >> 1)

x = 11
y = -2

# print bitwise left shift operator
print("a << 1 =", x << 1)
print("b << 1 =", y << 1)


```
