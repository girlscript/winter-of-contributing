## Python official website :  https://www.python.org/




# **Downloading Python :** 
* For downloading Python for **Mac-OS** https://www.python.org/downloads/macos/ .
* For downloading Python for **Windows** https://www.python.org/downloads/windows/ .
* For downloading any other version of python you can go to this https://www.python.org/downloads/ and choose your version.

# **Installing Python on your device :** 
In this example we are installing Python version 3.9.7 in Windows -

1. **Run the Python Installer** once downloaded.
2. Make sure you select the **"Install launcher for all users"** and **"Add Python 3.9 to PATH"** checkboxes.
py.jpeg

3. **Wait** for it to complete the installation process.
4. After it's completion, **Python is successfully installed** in your system.
5. Now for checking which version is installed in your system go to -
> Windows Search >>> Type "cmd" >>> Open it >> Type "python --version"
Screenshot 2021-09-27 213944.png

6. Kudos! python has been installed in your system, Now go to - 
> Windows Search >>> Type "IDLE" >>> Open it >>>  Start Coding
