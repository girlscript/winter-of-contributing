# What is the Tower of Hanoi?

Tower of Hanoi is a puzzle in which there are three rods and three disks(generally). The disks are of different sizes.
Tower of Hanoi, mathematical puzzle was invented by the French mathematician Edouard Lucas in the year 1883.
The aim is to transfer the disks from one source rod to another target or destination rod.

### Rules of solving the Puzzle

There are some rules which must not be violated while solving the puzzle. These rules are quite simple unlike the implementation and solution. But worry not, we will get through it.
There are three rods. The disks are arranged in the descending order i.e., the largest disk is placed at the bottom and the smallest one on the top. The rules are as follows:-

1. Only one disk can be moved among the towers at any given time.
2. Only the "top" disk can be removed.
3. No large disk can sit over a small disk.

#### The number of moves can be calculated using the formula 2^n-1 

We need to solve the puzzle in 7 moves from the above formula.
Suppose that the leftmost rod is called SOURCE, the rightmost rod is called TARGET and the rod in the middle is referred to as an AUX.


*Recursion- It is a technique in which a fuction/method calls itself until a specified condition is met.*

### Python Program to solve the Tower of Hanoi puzzle using recursion

```python
def TOH(ndisks, start, auxiliary, end):  #TOH stands for Tower of Hanoi, this is the recursive function.
    if(ndisks == 1):  
        print('Move disk 1 from rod {} to rod {}'.format(start, end))  
        #The format() method formats the specified values and inserts them inside the string's placeholder i.e., curly braces.
        return
    TOH(ndisks - 1, start, end, auxiliary)  #function calls itself
    print('Move disk {} from rod {} to rod {}'.format(ndisks, start, end))  
    TOH(ndisks - 1, auxiliary, start, end)  
  #Driver code
ndisks = int(input('Enter number of disks: '))
TOH(ndisks, 'A', 'B', 'C')
```

In the above solution, we have done the following things:-
    1. We have created a recursive function named TOH, and passed two arguments: the number of disks ndisks and the name of the rods such as start, auxiliary, and end.
    2. We can define the base case of TOH when the number of disks is 1. In this case, simply move the one disk from the source to target and return.
    3. Now, move remaining ndisks-1 disks from start to auxiliary using the target as the auxiliary.
    4. Then, the remaining 1 disk move on the start to end.
    5. Move the ndisks-1 disks on the auxiliary to the end using the start as the auxiliary.
